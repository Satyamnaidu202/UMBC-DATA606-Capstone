## 3. Data Cleaning & Preprocessing

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)
print('Columns with missing values:')
print(missing_df)

fig = px.bar(
    missing_df.reset_index(),
    x='index', y='Missing %',
    title='Missing Value Percentage by Column',
    labels={'index': 'Column', 'Missing %': 'Missing (%)'},
    color='Missing %', color_continuous_scale='reds'
)
fig.show()
```

```python
duplicates = df.duplicated().sum()
print(f'Duplicate rows detected: {duplicates:,} ({duplicates/len(df)*100:.1f}% of data)')
print('  These will be removed — they are likely system re-entries, not real distinct bookings.')
```

```python
df_clean = df.copy()
initial_rows = len(df_clean)

# Step 1: Drop exact duplicate rows
df_clean.drop_duplicates(inplace=True)
print(f'Step 1 – Removed duplicates        : -{initial_rows - len(df_clean):,} rows (now {len(df_clean):,})')

# Step 2: Remove bookings with 0 guests
before = len(df_clean)
df_clean = df_clean[
    (df_clean['adults'] + df_clean['children'].fillna(0) + df_clean['babies']) > 0
]
print(f'Step 2 – Removed 0-guest bookings   : -{before - len(df_clean):,} rows (now {len(df_clean):,})')

# Step 3: Remove negative ADR (impossible price)
before = len(df_clean)
df_clean = df_clean[df_clean['adr'] >= 0]
print(f'Step 3 – Removed negative ADR rows  : -{before - len(df_clean):,} rows (now {len(df_clean):,})')

# Step 4: Drop company column (94% missing — not usable)
df_clean.drop(columns=['company'], inplace=True)

# Step 5: Drop data-leakage columns
#   reservation_status      → directly encodes cancellation AFTER the fact
#   reservation_status_date → date of that outcome
#   assigned_room_type      → determined at check-in (post-booking), NOT at booking time
#                             Keeping reserved_room_type for features is fine
df_clean.drop(columns=['reservation_status', 'reservation_status_date'], inplace=True)
# NOTE: assigned_room_type is kept for EDA but will NOT be used as a model feature
# See Section 5 (Data Leak Audit) for full explanation
print('Step 4/5 – Dropped company + 2 leakage columns')

# Step 6: Impute remaining missing values
df_clean['children'].fillna(0, inplace=True)        # 4 rows — assume no children
df_clean['country'].fillna('Unknown', inplace=True) # 488 rows
df_clean['agent'].fillna(0, inplace=True)           # 16,340 rows — no agent

# Step 7: Cap ADR outliers at 99th percentile before ML (max=$5,400 is extreme)
adr_cap = df_clean['adr'].quantile(0.99)
before_cap = (df_clean['adr'] > adr_cap).sum()
df_clean['adr'] = df_clean['adr'].clip(upper=adr_cap)
print(f'Step 7 – Capped {before_cap} ADR outliers above ${adr_cap:.0f} (99th pct)')

STATS['clean_rows']       = len(df_clean)
STATS['removed_rows']     = initial_rows - len(df_clean)
STATS['cancel_rate_clean'] = df_clean['is_canceled'].mean() * 100
STATS['adr_cap']          = adr_cap

print(f'\n✅ Clean dataset: {len(df_clean):,} rows × {df_clean.shape[1]} columns')
print(f'   Removed {initial_rows - len(df_clean):,} rows ({(initial_rows-len(df_clean))/initial_rows*100:.1f}% of raw)')
print(f'   Remaining missing values: {df_clean.isnull().sum().sum()}')
print(f'   Cancellation rate after cleaning: {STATS["cancel_rate_clean"]:.1f}%')
print(f'   ADR capped at ${adr_cap:.2f} (99th pct) — affects {before_cap} rows')
```

## 4. Exploratory Data Analysis (EDA)
> **Note:** All EDA is performed on the cleaned dataset. All percentage figures cited in insights are computed dynamically from actual data. `required_car_parking_spaces` and `assigned_room_type` are included in EDA for documentation but will NOT be used as model features (see Section 5).

### 4.1 Cancellation by Hotel Type

```python
hotel_cancel = df_clean.groupby('hotel')['is_canceled'].agg(['sum','count'])
hotel_cancel['cancel_rate'] = (hotel_cancel['sum']/hotel_cancel['count']*100).round(1)
hotel_cancel = hotel_cancel.reset_index()

city_rate   = hotel_cancel.loc[hotel_cancel['hotel']=='City Hotel',   'cancel_rate'].values[0]
resort_rate = hotel_cancel.loc[hotel_cancel['hotel']=='Resort Hotel', 'cancel_rate'].values[0]
STATS['city_cancel_rate']   = city_rate
STATS['resort_cancel_rate'] = resort_rate

print(hotel_cancel.to_string(index=False))
print(f'\n💡 Insight: City Hotels cancel at {city_rate}% vs Resort Hotels at {resort_rate}%.')
print('   City Hotels attract more transient business travellers who book flexible rates.')

fig = px.bar(
    hotel_cancel, x='hotel', y='cancel_rate', text='cancel_rate', color='hotel',
    title='Cancellation Rate by Hotel Type (%)',
    labels={'cancel_rate': 'Cancellation Rate (%)', 'hotel': 'Hotel Type'},
    color_discrete_sequence=['#3498db', '#e74c3c']
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

### 4.2 Lead Time Distribution vs Cancellation

```python
lead_by_outcome = df_clean.groupby('is_canceled')['lead_time'].mean().round(1)
lt_not_canceled = lead_by_outcome[0]
lt_canceled     = lead_by_outcome[1]
lt_pct_longer   = round((lt_canceled - lt_not_canceled) / lt_not_canceled * 100, 1)
STATS['lt_not_canceled'] = lt_not_canceled
STATS['lt_canceled']     = lt_canceled
STATS['lt_pct_longer']   = lt_pct_longer

print(f'Average lead time — Not Canceled: {lt_not_canceled} days | Canceled: {lt_canceled} days')
print(f'💡 Insight: Cancelled bookings have {lt_pct_longer}% longer lead time.')
print('   The further ahead someone books, the more likely their plans change.')

fig = px.box(
    df_clean, x='is_canceled', y='lead_time', color='is_canceled',
    title='Lead Time Distribution: Canceled vs Not Canceled',
    labels={'is_canceled': 'Canceled (1=Yes)', 'lead_time': 'Lead Time (days)'},
    color_discrete_map={0: '#2ecc71', 1: '#e74c3c'}
)
fig.show()
```

### 4.3 Deposit Type — Key Anomaly

> ⚠️ **Important Nuance:** Non Refund bookings have a very high cancellation rate. In this dataset, non-refundable bookings are often recorded as "Canceled" even when the guest forfeits the deposit and the hotel keeps the money. The hotel still receives revenue — this is a data recording artefact, not a pure business cancellation signal. After deduplication this rate has dropped from ~99% to ~95%, confirming the duplicates were inflating it further.

```python
deposit_cancel = df_clean.groupby('deposit_type')['is_canceled'].agg(['sum','count'])
deposit_cancel['cancel_rate'] = (deposit_cancel['sum']/deposit_cancel['count']*100).round(1)
deposit_cancel = deposit_cancel.reset_index()

non_refund_rate = deposit_cancel.loc[deposit_cancel['deposit_type']=='Non Refund','cancel_rate'].values[0]
STATS['non_refund_cancel_rate'] = non_refund_rate

print(deposit_cancel[['deposit_type','count','sum','cancel_rate']].to_string(index=False))
print(f'\n💡 Insight: Non Refund deposit type = {non_refund_rate}% cancel rate.')
print('   This is a dataset artefact — hotel system records these as canceled even when')
print('   revenue is retained. Use with caution in model interpretation.')

fig = px.bar(
    deposit_cancel, x='deposit_type', y='cancel_rate', text='cancel_rate', color='deposit_type',
    title='Cancellation Rate by Deposit Type (%)',
    labels={'cancel_rate': 'Cancellation Rate (%)', 'deposit_type': 'Deposit Type'}
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

### 4.4 Cancellation by Market Segment

```python
market_cancel = df_clean.groupby('market_segment')['is_canceled'].agg(['sum','count'])
market_cancel['cancel_rate'] = (market_cancel['sum']/market_cancel['count']*100).round(1)
market_cancel = market_cancel.sort_values('cancel_rate', ascending=False).reset_index()

market_for_insight = market_cancel[market_cancel['market_segment'] != 'Undefined']
top_segment    = market_for_insight.iloc[0]
second_segment = market_for_insight.iloc[1]
safest_segment = market_for_insight.iloc[-1]
STATS['top_market']    = f"{top_segment['market_segment']} ({top_segment['cancel_rate']}%)"
STATS['safest_market'] = f"{safest_segment['market_segment']} ({safest_segment['cancel_rate']}%)"

print(market_cancel.to_string(index=False))
print(f"\n💡 Insight: Highest-risk segments: {top_segment['market_segment']} ({top_segment['cancel_rate']}%), "
      f"{second_segment['market_segment']} ({second_segment['cancel_rate']}%)")
print(f"   Safest segment: {safest_segment['market_segment']} ({safest_segment['cancel_rate']}%)")
print("   Note: 'Undefined' segment at 100% has very few bookings — a data label issue, not a true insight.")

fig = px.bar(
    market_cancel, x='market_segment', y='cancel_rate', text='cancel_rate',
    color='cancel_rate',
    title='Cancellation Rate by Market Segment (%)',
    labels={'cancel_rate': 'Cancellation Rate (%)', 'market_segment': 'Market Segment'},
    color_continuous_scale='RdYlGn_r'
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

### 4.6 Monthly Cancellation Trend

```python
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly = df_clean.groupby('arrival_date_month')['is_canceled'].agg(['sum','count'])
monthly['cancel_rate'] = (monthly['sum']/monthly['count']*100).round(1)
monthly = monthly.reindex(month_order).reset_index()

peak_idx   = monthly['cancel_rate'].idxmax()
low_idx    = monthly['cancel_rate'].idxmin()
peak_month = monthly.loc[peak_idx, 'arrival_date_month']
peak_rate  = monthly.loc[peak_idx, 'cancel_rate']
low_month  = monthly.loc[low_idx, 'arrival_date_month']
low_rate   = monthly.loc[low_idx, 'cancel_rate']
STATS['peak_month'] = peak_month
STATS['peak_rate']  = peak_rate
STATS['low_month']  = low_month
STATS['low_rate']   = low_rate

print(f'Peak cancellation  : {peak_month} ({peak_rate}%)')
print(f'Lowest cancellation: {low_month} ({low_rate}%)')
print(f'\n💡 Insight: Cancellation risk peaks in {peak_month} ({peak_rate}%) and is lowest in {low_month} ({low_rate}%).')

fig = px.line(
    monthly, x='arrival_date_month', y='cancel_rate', markers=True,
    title=f'Monthly Cancellation Rate (%) — Peak: {peak_month}, Lowest: {low_month}',
    labels={'arrival_date_month': 'Month', 'cancel_rate': 'Cancellation Rate (%)'}
)
fig.update_traces(line_color='#e74c3c', marker_color='#c0392b', marker_size=8)
fig.show()
```

```python
parking_full = df_clean.groupby('required_car_parking_spaces')['is_canceled'].agg(['mean','count'])
parking_full['cancel_rate'] = (parking_full['mean'] * 100).round(1)
parking_full = parking_full.reset_index()
parking_full.columns = ['required_car_parking_spaces','mean','n_bookings','cancel_rate']
parking_full = parking_full[parking_full['required_car_parking_spaces'] <= 3]

zero_parking_cancel = parking_full.loc[parking_full['required_car_parking_spaces']==0,'cancel_rate'].values[0]

print('Car parking spaces breakdown (with volume):')
print(parking_full[['required_car_parking_spaces','n_bookings','cancel_rate']].to_string(index=False))
print(f'\n🚫 0% cancellation for 1+ spaces across {parking_full[parking_full["required_car_parking_spaces"]>=1]["n_bookings"].sum():,} bookings is IMPOSSIBLE for real data.')
print('   This is a deduplication artifact. This feature will NOT be used in the ML model.')
print('   The directional insight (parking = more planning = lower cancel) is real but the magnitude is not.')

fig = px.bar(
    parking_full, x='required_car_parking_spaces', y='cancel_rate',
    text=parking_full['cancel_rate'].round(1),
    title='Cancellation Rate by Car Parking Spaces Requested<br><sup>⚠️ EDA ONLY — 0% for 1+ spaces is a data artifact — feature excluded from model</sup>',
    labels={'required_car_parking_spaces': 'Parking Spaces Requested', 'cancel_rate': 'Cancellation Rate (%)'},
    color='cancel_rate', color_continuous_scale='RdYlGn_r'
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

## 5. Data Leak & Feature Audit
> This is the most critical section. Before building any model, every candidate feature must pass a temporal validity test: **Is this information available at the moment of prediction (booking creation time)?** Features derived from post-booking events are data leaks.

### Features Audited for Leakage

```python
print('='*70)
print('DATA LEAK AUDIT — ALL CANDIDATE FEATURES')
print('='*70)

audit = [
    ('reservation_status',       'DROPPED', 'Directly encodes cancellation outcome — perfect leak'),
    ('reservation_status_date',  'DROPPED', 'Post-event date — perfect leak'),
    ('assigned_room_type',       'DROPPED from MODEL', 'Set at check-in, NOT at booking time (see INVESTIGATION below)'),
    ('room_type_changed',        'EXCLUDED', 'Derived from assigned_room_type — inherits the leak'),
    ('required_car_parking_spaces','EXCLUDED', '0% artifact corrupts model; directional EDA insight retained'),
    ('deposit_type',             'INCLUDED (with analysis)', 'Known at booking; recording artifact documented in Sec 9'),
    ('lead_time',                'INCLUDED', 'Known at booking creation time'),
    ('adr',                      'INCLUDED (capped)', 'Rate agreed at booking time; outliers capped'),
    ('booking_changes',          'INCLUDED*', 'Dynamic: updates as booking matures; 0 at booking creation'),
    ('market_segment',           'INCLUDED', 'Known at booking creation time'),
    ('total_of_special_requests','INCLUDED', 'Known at booking creation time'),
    ('previous_cancellations',   'INCLUDED', 'Historical — known at booking time'),
    ('agent',                    'INCLUDED', 'Known at booking creation time'),
    ('customer_type',            'INCLUDED', 'Known at booking creation time'),
    ('meal',                     'INCLUDED', 'Selected at booking time'),
    ('distribution_channel',     'INCLUDED', 'Known at booking creation time'),
    ('hotel',                    'INCLUDED', 'Known at booking creation time'),
    ('is_repeated_guest',        'INCLUDED', 'Historical flag — known at booking time'),
]

print(f'{"Feature":<35} {"Status":<25} {"Reason"}')
print('-'*100)
for feat, status, reason in audit:
    flag = '🔴' if 'DROPPED' in status or 'EXCLUDED' in status else '✅'
    print(f'{flag} {feat:<33} {status:<25} {reason}')

print('\n* booking_changes = 0 at booking creation; valid for dynamic prediction updates over booking lifetime')
```

### 5.1 Deep Investigation: `assigned_room_type` / `room_type_changed`

```python
print('INVESTIGATION: Is room_type_changed a data leak?')
print('='*55)
print('Hypothesis: assigned_room_type is determined at CHECK-IN.')
print('  → Canceled bookings never check in.')
print('  → Canceled bookings should have assigned == reserved (no change).')
print('  → room_type_changed=1 would ONLY appear in non-canceled bookings.')
print('  → This makes it a near-perfect proxy for "not canceled" = DATA LEAK.\n')

# Compute room_type_changed in df_clean (which still has assigned_room_type)
df_clean['_room_type_changed'] = (df_clean['reserved_room_type'] != df_clean['assigned_room_type']).astype(int)

rtc_cancel = df_clean.groupby('_room_type_changed')['is_canceled'].agg(['mean','count'])
rtc_cancel['cancel_rate'] = (rtc_cancel['mean'] * 100).round(2)
rtc_cancel = rtc_cancel.reset_index()

print('Cancel rate by room_type_changed:')
print(rtc_cancel[['_room_type_changed','count','cancel_rate']].rename(
    columns={'_room_type_changed':'room_type_changed', 'count':'n_bookings'}
).to_string(index=False))

rate_changed    = rtc_cancel.loc[rtc_cancel['_room_type_changed']==1, 'cancel_rate'].values[0]
rate_unchanged  = rtc_cancel.loc[rtc_cancel['_room_type_changed']==0, 'cancel_rate'].values[0]

print(f'\n  Room NOT changed → {rate_unchanged:.1f}% cancel rate')
print(f'  Room IS changed  → {rate_changed:.1f}% cancel rate')

if rate_changed < 5:
    print('\n🔴 CONFIRMED DATA LEAK: Near-0% cancellation for room_type_changed=1.')
    print('   If the room was changed (reassigned), that can only happen at check-in.')
    print('   Canceled bookings never reach check-in → this feature leaks the outcome.')
    print('   → room_type_changed and assigned_room_type EXCLUDED from all model features.')
else:
    print(f'\n⚠️  Inconclusive: rate_changed={rate_changed:.1f}%. Review manually.')
    print('   Room assignments may sometimes be pre-set before check-in in this system.')

# Clean up the temporary column
df_clean.drop(columns=['_room_type_changed'], inplace=True)
```

### 5.2 Deep Investigation: `required_car_parking_spaces` Artifact

```python
print('INVESTIGATION: Why is required_car_parking_spaces = 0% for parking>0?')
print('='*60)
print('This is a deduplication artifact:')
print('  → Before dedup: 119,390 rows — some canceled bookings with parking existed')
print('  → After dedup:   87,229 rows — the 31,994 removed duplicates disproportionately')
print('    contained canceled bookings with parking spaces')
print('  → Result: ALL remaining parking>0 bookings are non-canceled in the clean set')
print()

# Verify in raw data
raw_parking = df[df['required_car_parking_spaces'] >= 1].copy()
raw_parking_cancel = raw_parking['is_canceled'].mean() * 100
raw_parking_n = len(raw_parking)

clean_parking = df_clean[df_clean['required_car_parking_spaces'] >= 1].copy()
clean_parking_cancel = clean_parking['is_canceled'].mean() * 100
clean_parking_n = len(clean_parking)

print(f'Raw data   — bookings with parking: {raw_parking_n:,} | cancel rate: {raw_parking_cancel:.1f}%')
print(f'Clean data — bookings with parking: {clean_parking_n:,} | cancel rate: {clean_parking_cancel:.1f}%')
print(f'\n  The raw cancel rate ({raw_parking_cancel:.1f}%) is realistic.')
print(f'  The post-dedup rate ({clean_parking_cancel:.1f}%) is an artifact of which duplicates were removed.')
print('\n🚫 DECISION: required_car_parking_spaces EXCLUDED from model.')
print('   Model trained on this would learn: parking=1 → never cancels.')
print('   This will FAIL to generalize to new data where that is untrue.')
```

## 6. Feature Engineering & ML Preparation
> **Corrected from v3:** `room_type_changed` removed (data leak). `required_car_parking_spaces` removed (artifact). `lead_time_bucket` now uses `OrdinalEncoder` (correct for ordered categories). ADR already capped in Section 3.

```python
df_ml = df_clean.copy()

# --- Derived features — all valid at booking time (no leakage) ---
df_ml['total_nights']    = df_ml['stays_in_weekend_nights'] + df_ml['stays_in_week_nights']
df_ml['total_guests']    = df_ml['adults'] + df_ml['children'] + df_ml['babies']
df_ml['has_agent']       = (df_ml['agent'] > 0).astype(int)
df_ml['is_weekend_only'] = (
    (df_ml['stays_in_weekend_nights'] > 0) & (df_ml['stays_in_week_nights'] == 0)
).astype(int)

# REMOVED: room_type_changed — DATA LEAK (uses assigned_room_type which is set at check-in)

# lead_time_bucket — ordered categories, use OrdinalEncoder not LabelEncoder
lead_bins   = [-1, 7, 30, 90, 180, 365, 9999]
lead_labels = ['0-7d','8-30d','31-90d','91-180d','181-365d','365+d']
df_ml['lead_time_bucket'] = pd.cut(
    df_ml['lead_time'],
    bins=lead_bins,
    labels=lead_labels
)

print('Derived feature distributions (sanity check):')
print(f'  has_agent        : {df_ml["has_agent"].value_counts().to_dict()}')
print(f'  is_weekend_only  : {df_ml["is_weekend_only"].value_counts().to_dict()}')
print(f'  lead_time_bucket : {df_ml["lead_time_bucket"].value_counts().sort_index().to_dict()}')
print()
print('🚫 room_type_changed NOT created — confirmed data leak (see Section 5.1)')
print('🚫 required_car_parking_spaces NOT in feature list — artifact (see Section 5.2)')
```

```python
# PRIMARY feature set (includes deposit_type)
# REMOVED vs v3: room_type_changed, required_car_parking_spaces
features_full = [
    'lead_time', 'total_nights', 'total_guests', 'adr',
    'previous_cancellations', 'previous_bookings_not_canceled',
    'booking_changes', 'days_in_waiting_list',
    'total_of_special_requests',
    'is_repeated_guest', 'has_agent', 'is_weekend_only',
    'hotel', 'meal', 'market_segment', 'distribution_channel',
    'deposit_type', 'customer_type', 'lead_time_bucket'
]

# SECONDARY — without deposit_type (to verify model doesn't over-rely on it)
features_no_deposit = [f for f in features_full if f != 'deposit_type']

target       = 'is_canceled'
nom_cat_cols = ['hotel', 'meal', 'market_segment', 'distribution_channel',
                'deposit_type', 'customer_type']
ord_cat_cols = ['lead_time_bucket']  # ordered — use OrdinalEncoder

lead_time_bucket_order = ['0-7d','8-30d','31-90d','91-180d','181-365d','365+d']

def prepare_X(df_in, feature_list):
    """Encode features: OrdinalEncoder for lead_time_bucket, LabelEncoder for nominals."""
    df_out = df_in[feature_list].copy()
    # Ordinal encoding for lead_time_bucket (preserves order)
    if 'lead_time_bucket' in df_out.columns:
        oe = OrdinalEncoder(categories=[lead_time_bucket_order])
        df_out['lead_time_bucket'] = oe.fit_transform(
            df_out[['lead_time_bucket']].astype(str)
        ).flatten()
    # Label encoding for remaining categoricals
    le = LabelEncoder()
    for col in nom_cat_cols:
        if col in df_out.columns:
            df_out[col] = le.fit_transform(df_out[col].astype(str))
    return df_out

X_full       = prepare_X(df_ml, features_full)
X_no_deposit = prepare_X(df_ml, features_no_deposit)
y = df_ml[target]

print(f'Full feature set     : {X_full.shape[1]} features')
print(f'Without deposit_type : {X_no_deposit.shape[1]} features')
print(f'\nFeatures in full set : {list(X_full.columns)}')
```

```python
X_train_f, X_test_f, y_train, y_test = train_test_split(
    X_full, y, test_size=0.2, random_state=42, stratify=y
)
X_train_nd, X_test_nd, _, _ = train_test_split(
    X_no_deposit, y, test_size=0.2, random_state=42, stratify=y
)

STATS['train_rows'] = X_train_f.shape[0]
STATS['test_rows']  = X_test_f.shape[0]
STATS['n_features'] = X_full.shape[1]

print(f'Training rows : {X_train_f.shape[0]:,}')
print(f'Test rows     : {X_test_f.shape[0]:,}')
print(f'Features used : {X_full.shape[1]}')
print()
print('Class balance in test set:')
test_balance = y_test.value_counts(normalize=True)
print(f'  Not Canceled (0): {test_balance[0]:.1%}')
print(f'  Canceled     (1): {test_balance[1]:.1%}')
print()
print('💡 Class imbalance note: ~72.5% not-canceled vs ~27.5% canceled.')
print('   This is WHY recall at default threshold (0.5) will be lower than precision.')
print('   The model learns the majority class better. Threshold tuning (Section 8)')
print('   is the correct solution — NOT oversampling, which would inflate metrics.')
```

## 7. Model Training & Evaluation

```python
def evaluate_model(name, model, X_tr, X_te, y_tr, y_te):
    """Fit model and return metrics dict + artefacts."""
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_te)
    y_prob = model.predict_proba(X_te)[:, 1]
    return {
        'Model'    : name,
        'Accuracy' : accuracy_score(y_te, y_pred),
        'Precision': precision_score(y_te, y_pred),
        'Recall'   : recall_score(y_te, y_pred),
        'F1 Score' : f1_score(y_te, y_pred),
        'ROC-AUC'  : roc_auc_score(y_te, y_prob),
    }, model, y_pred, y_prob

all_results    = []
trained_models = {}

scaler     = StandardScaler()
X_train_sc = scaler.fit_transform(X_train_f)
X_test_sc  = scaler.transform(X_test_f)

# Logistic Regression — class_weight='balanced' to handle class imbalance
m, model_lr, yp_lr, ypr_lr = evaluate_model(
    'Logistic Regression',
    LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42),
    X_train_sc, X_test_sc, y_train, y_test
)
all_results.append(m)
trained_models['Logistic Regression'] = (model_lr, yp_lr, ypr_lr)
print(f"Logistic Regression — Accuracy: {m['Accuracy']:.4f} | ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
m, model_dt, yp_dt, ypr_dt = evaluate_model(
    'Decision Tree',
    DecisionTreeClassifier(max_depth=8, min_samples_leaf=50,
                           class_weight='balanced', random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m)
trained_models['Decision Tree'] = (model_dt, yp_dt, ypr_dt)
print(f"Decision Tree       — Accuracy: {m['Accuracy']:.4f} | ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
m, model_rf, yp_rf, ypr_rf = evaluate_model(
    'Random Forest',
    RandomForestClassifier(n_estimators=200, max_depth=12,
                           min_samples_leaf=20, class_weight='balanced',
                           n_jobs=-1, random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m)
trained_models['Random Forest'] = (model_rf, yp_rf, ypr_rf)
print(f"Random Forest       — Accuracy: {m['Accuracy']:.4f} | ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
m, model_gb, yp_gb, ypr_gb = evaluate_model(
    'Gradient Boosting',
    GradientBoostingClassifier(n_estimators=200, learning_rate=0.1,
                               max_depth=5, random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m)
trained_models['Gradient Boosting'] = (model_gb, yp_gb, ypr_gb)
m_gb = m
print(f"Gradient Boosting   — Accuracy: {m['Accuracy']:.4f} | ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
if XGBOOST_AVAILABLE:
    m, model_xgb, yp_xgb, ypr_xgb = evaluate_model(
        'XGBoost',
        XGBClassifier(n_estimators=200, learning_rate=0.1, max_depth=6,
                      scale_pos_weight=(y_train==0).sum()/(y_train==1).sum(),
                      eval_metric='logloss', n_jobs=-1, random_state=42),
        X_train_f, X_test_f, y_train, y_test
    )
    all_results.append(m)
    trained_models['XGBoost'] = (model_xgb, yp_xgb, ypr_xgb)
    print(f"XGBoost             — Accuracy: {m['Accuracy']:.4f} | ROC-AUC: {m['ROC-AUC']:.4f}")
else:
    print('XGBoost skipped (not installed). Install with: pip install xgboost')
    print('XGBoost typically achieves ROC-AUC ~0.01-0.02 higher than Gradient Boosting.')
```

```python
results_df = pd.DataFrame(all_results).set_index('Model')
results_df = results_df.sort_values('ROC-AUC', ascending=False)

best_model_name = results_df.index[0]
STATS['best_model']    = best_model_name
STATS['best_roc_auc']  = results_df['ROC-AUC'].iloc[0]
STATS['best_accuracy'] = results_df['Accuracy'].iloc[0]
STATS['best_recall']   = results_df['Recall'].iloc[0]
STATS['best_f1']       = results_df['F1 Score'].iloc[0]

print('Model Comparison — Corrected Feature Set (v4):')
print('(room_type_changed and required_car_parking_spaces excluded)')
print()
print(results_df.round(4).to_string())

print('\n📝 Performance vs v3:')
print('  v3 Gradient Boosting ROC-AUC: 0.8605 (included 2 leaky/artifact features)')
print(f'  v4 {best_model_name} ROC-AUC: {STATS["best_roc_auc"]:.4f} (clean, trustworthy features)')
print('  The gap represents how much the model was relying on invalid signals.')
print('  A lower but HONEST metric is always preferable to an inflated one.')
```

## 8. Threshold Tuning — Business-Relevant Recall vs Precision
> Default threshold = 0.5. For hotels, catching cancellations early (high Recall) is usually more valuable than precision. Lowering the threshold catches more cancellations but also flags some genuine stays — hotels must decide their tolerance.

```python
best_prob = trained_models[best_model_name][2]
print(f'Threshold tuning on: {best_model_name}\n')

thresholds     = np.arange(0.20, 0.75, 0.05)
threshold_rows = []

for thr in thresholds:
    y_pred_thr = (best_prob >= thr).astype(int)
    threshold_rows.append({
        'Threshold' : round(thr, 2),
        'Accuracy'  : accuracy_score(y_test, y_pred_thr),
        'Precision' : precision_score(y_test, y_pred_thr, zero_division=0),
        'Recall'    : recall_score(y_test, y_pred_thr, zero_division=0),
        'F1 Score'  : f1_score(y_test, y_pred_thr, zero_division=0),
    })

thr_df = pd.DataFrame(threshold_rows)
print(thr_df.round(3).to_string(index=False))

best_f1_thr = thr_df.loc[thr_df['F1 Score'].idxmax(), 'Threshold']
STATS['recommended_threshold'] = best_f1_thr
print(f'\n📌 Threshold with best F1 Score: {best_f1_thr}')
print('   Use 0.30-0.35 for high-recall (hotel wants to catch as many cancellations as possible)')
print('   Use 0.50+ for high-precision (hotel wants fewer false alarms)')
```

```python
fig = go.Figure()
for metric, color in [('Recall','#e74c3c'),('Precision','#3498db'),
                       ('F1 Score','#2ecc71'),('Accuracy','#f39c12')]:
    fig.add_trace(go.Scatter(
        x=thr_df['Threshold'], y=thr_df[metric],
        mode='lines+markers', name=metric, line=dict(color=color, width=2)
    ))

fig.add_vline(x=0.5, line_dash='dash', line_color='gray',
              annotation_text='Default (0.5)', annotation_position='top right')
fig.add_vline(x=best_f1_thr, line_dash='dot', line_color='red',
              annotation_text=f'Best F1 ({best_f1_thr})', annotation_position='top left')

fig.update_layout(
    title=f'Threshold Tuning — {best_model_name}',
    xaxis_title='Decision Threshold',
    yaxis_title='Score', height=450
)
fig.show()

recommended_thr    = best_f1_thr
y_pred_recommended = (best_prob >= recommended_thr).astype(int)
print(f'\nAt threshold = {recommended_thr} (best-F1 / recommended mode):')
print(classification_report(y_test, y_pred_recommended,
                             target_names=['Not Canceled', 'Canceled']))
```

```python
fi_nd = pd.DataFrame({
    'Feature'   : X_no_deposit.columns,
    'Importance': model_gb_nd.feature_importances_
}).sort_values('Importance', ascending=False)

fig = px.bar(
    fi_nd, x='Importance', y='Feature', orientation='h',
    title='Feature Importances – Gradient Boosting (WITHOUT deposit_type)',
    color='Importance', color_continuous_scale='Oranges_r'
)
fig.update_layout(yaxis={'categoryorder':'total ascending'}, height=600)
fig.show()

STATS['top_features_no_deposit'] = fi_nd['Feature'].head(6).tolist()

print('Feature rankings WITHOUT deposit_type (true business drivers):')
print(fi_nd.to_string(index=False))
```

```python
print(f'Running 5-Fold Stratified Cross-Validation on {best_model_name}...')
cv_model  = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
skf       = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(cv_model, X_full, y, cv=skf, scoring='roc_auc', n_jobs=-1)

STATS['cv_mean'] = cv_scores.mean()
STATS['cv_std']  = cv_scores.std()

print(f'\nROC-AUC per fold : {cv_scores.round(4)}')
print(f'Mean ROC-AUC     : {cv_scores.mean():.4f}')
print(f'Std Dev          : {cv_scores.std():.4f}')

if cv_scores.std() < 0.01:
    print(f'✅ Std dev {cv_scores.std():.4f} < 0.01 — model is stable, no overfitting.')
elif cv_scores.std() < 0.02:
    print(f'✅ Std dev {cv_scores.std():.4f} — model is reasonably stable.')
else:
    print(f'⚠️  Std dev {cv_scores.std():.4f} — some variance across folds, consider regularization.')

gap = STATS['best_roc_auc'] - cv_scores.mean()
print(f'\nTest AUC vs CV mean: {STATS["best_roc_auc"]:.4f} vs {cv_scores.mean():.4f} (gap={gap:.4f})')
if abs(gap) < 0.01:
    print('✅ Gap < 0.01 — test set result is consistent with cross-validation. No lucky split.')
else:
    print('⚠️  Non-trivial gap — review train/test split randomness.')

fig = px.bar(
    x=[f'Fold {i+1}' for i in range(5)], y=cv_scores,
    text=cv_scores.round(4),
    title=f'5-Fold CV ROC-AUC — {best_model_name} (v4)',
    labels={'x':'Fold','y':'ROC-AUC'},
    color=cv_scores, color_continuous_scale='Blues'
)
fig.add_hline(y=cv_scores.mean(), line_dash='dash', line_color='red',
              annotation_text=f'Mean={cv_scores.mean():.4f}')
fig.update_traces(texttemplate='%{text}', textposition='outside')
fig.show()
```

## 11. SHAP Value Analysis — Model Explainability
> SHAP (SHapley Additive exPlanations) tells us **why** the model made each prediction. This is critical for the Streamlit app (per-booking explanations) and for business trust in the model.
> 
> Install if needed: `pip install shap`

```python
if SHAP_AVAILABLE:
    print('Computing SHAP values for Gradient Boosting model...')
    print('(Using 2000-row sample for speed — full dataset would take longer)')

    # Use a sample for speed
    sample_idx = np.random.RandomState(42).choice(len(X_test_f), size=2000, replace=False)
    X_shap = X_test_f.iloc[sample_idx]

    explainer   = shap.TreeExplainer(model_gb)
    shap_values = explainer.shap_values(X_shap)

    print('✅ SHAP values computed.')
    print(f'   Shape: {shap_values.shape}')
    print()

    # Mean absolute SHAP values — global feature importance
    mean_abs_shap = pd.DataFrame({
        'Feature'        : X_shap.columns,
        'Mean |SHAP|'    : np.abs(shap_values).mean(axis=0)
    }).sort_values('Mean |SHAP|', ascending=False)

    print('Global feature importance via mean |SHAP|:')
    print(mean_abs_shap.to_string(index=False))
    STATS['shap_top_feature'] = mean_abs_shap.iloc[0]['Feature']

else:
    print('⚠️  SHAP not available. Install with: pip install shap')
    print('   Skipping SHAP analysis. Run after installation.')
```

```python
if SHAP_AVAILABLE:
    # Summary plot — dot plot showing feature impact distribution
    print('SHAP Summary Plot (Beeswarm):')
    print('Red dots = higher feature value, Blue = lower value')
    print('Right of center = pushes toward cancellation, Left = pushes against it')
    plt.figure(figsize=(10, 8))
    shap.summary_plot(shap_values, X_shap, show=True, max_display=15)
else:
    print('SHAP not available — skipping summary plot.')
```

```python
if SHAP_AVAILABLE:
    # Bar plot of mean absolute SHAP values
    print('SHAP Bar Plot — Global Feature Importance:')
    plt.figure(figsize=(10, 6))
    shap.summary_plot(shap_values, X_shap, plot_type='bar', show=True, max_display=15)
else:
    print('SHAP not available — skipping bar plot.')
```

```python
if SHAP_AVAILABLE:
    # Single prediction explanation (Force plot saved as static)
    print('SHAP Force Plot — Example Prediction Explanation:')
    print('This shows WHY the model assigned a specific probability to one booking.')
    print()

    # Pick a high-risk booking (predicted cancel probability > 0.6)
    best_proba     = trained_models[best_model_name][2]
    best_proba_sample = best_proba[sample_idx]
    high_risk_idx  = np.where(best_proba_sample > 0.6)[0]

    if len(high_risk_idx) > 0:
        idx = high_risk_idx[0]
        print(f'Example booking (high-risk, pred prob={best_proba_sample[idx]:.3f}):')
        print(X_shap.iloc[idx].to_dict())
        print()
        print('SHAP values for this booking (positive = pushes toward cancellation):')
        booking_shap = pd.DataFrame({
            'Feature': X_shap.columns,
            'Feature Value': X_shap.iloc[idx].values,
            'SHAP Value': shap_values[idx]
        }).sort_values('SHAP Value', key=abs, ascending=False)
        print(booking_shap.to_string(index=False))

        # Waterfall plot
        plt.figure(figsize=(10, 6))
        explanation = shap.Explanation(
            values=shap_values[idx],
            base_values=explainer.expected_value,
            data=X_shap.iloc[idx].values,
            feature_names=list(X_shap.columns)
        )
        shap.waterfall_plot(explanation, max_display=10, show=True)
    else:
        print('No high-risk bookings found in sample. Showing first booking.')
        shap.waterfall_plot(
            shap.Explanation(
                values=shap_values[0],
                base_values=explainer.expected_value,
                data=X_shap.iloc[0].values,
                feature_names=list(X_shap.columns)
            ), max_display=10, show=True
        )
else:
    print('SHAP not available — skipping force plot.')
```
