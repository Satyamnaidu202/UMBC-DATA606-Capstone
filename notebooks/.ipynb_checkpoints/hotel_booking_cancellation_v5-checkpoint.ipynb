## 3. Data Cleaning & Preprocessing

```python
# ── 3.1  Missing values ────────────────────────────────────────────────────────
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
miss_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
miss_df = miss_df[miss_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)
print('Columns with missing values:')
print(miss_df.to_string())

fig = px.bar(
    miss_df.reset_index(), x='index', y='Missing %',
    title='Missing Value Percentage by Column',
    labels={'index': 'Column', 'Missing %': 'Missing (%)'},
    color='Missing %', color_continuous_scale='Reds'
)
fig.show()
```

```python
# ── 3.2  Duplicates ────────────────────────────────────────────────────────────
dupes = df.duplicated().sum()
print(f'Exact duplicate rows: {dupes:,}  ({dupes / len(df) * 100:.1f}% of raw data)')
print()
print('These are system re-entries — not distinct bookings.')
print('Critically: duplicates are disproportionately canceled bookings,')
print('which is why the overall cancellation rate drops substantially after removal.')

# Show that duplicates are more likely to be canceled
dup_mask = df.duplicated(keep=False)  # mark ALL copies
print(f'\nCancellation rate among duplicated rows : '
      f'{df.loc[dup_mask, "is_canceled"].mean()*100:.1f}%')
print(f'Cancellation rate among non-dup rows    : '
      f'{df.loc[~dup_mask, "is_canceled"].mean()*100:.1f}%')
```

```python
# ── 3.3  Full cleaning pipeline ────────────────────────────────────────────────
df_clean = df.copy()
initial_rows = len(df_clean)

# Step 1 — Remove exact duplicates
df_clean.drop_duplicates(inplace=True)
print(f'Step 1 – Removed duplicates            : -{initial_rows - len(df_clean):,}  (now {len(df_clean):,})')

# Step 2 — Remove bookings with 0 guests (impossible)
before = len(df_clean)
df_clean = df_clean[
    (df_clean['adults'] + df_clean['children'].fillna(0) + df_clean['babies']) > 0
]
print(f'Step 2 – Removed 0-guest bookings      : -{before - len(df_clean):,}  (now {len(df_clean):,})')

# Step 3 — Remove negative ADR (impossible price)
before = len(df_clean)
df_clean = df_clean[df_clean['adr'] >= 0]
print(f'Step 3 – Removed negative ADR          : -{before - len(df_clean):,}  (now {len(df_clean):,})')

# Step 4 — Drop `company` (94% missing — not recoverable)
df_clean.drop(columns=['company'], inplace=True)

# Step 5 — Drop obvious data-leakage columns
#   reservation_status      → encodes the cancellation outcome directly
#   reservation_status_date → the date of that outcome
# NOTE: assigned_room_type is KEPT in df_clean for EDA & audit but will NOT be
#       used as a model feature (see Section 5 for full investigation).
df_clean.drop(columns=['reservation_status', 'reservation_status_date'], inplace=True)
print('Step 4/5 – Dropped company + 2 leakage columns (reservation_status*)')

# Step 6 — Impute small missing values
df_clean['children'].fillna(0, inplace=True)       # 4 rows
df_clean['country'].fillna('Unknown', inplace=True) # 488 rows
df_clean['agent'].fillna(0, inplace=True)           # 16,340 rows — 0 = no agent

# Step 7 — Cap ADR at 99th percentile (max of $5,400 is an extreme outlier)
adr_cap = df_clean['adr'].quantile(0.99)
n_capped = (df_clean['adr'] > adr_cap).sum()
df_clean['adr'] = df_clean['adr'].clip(upper=adr_cap)
print(f'Step 7 – Capped {n_capped} ADR outliers above ${adr_cap:.2f} (99th pct)')

STATS['clean_rows']        = len(df_clean)
STATS['removed_rows']      = initial_rows - len(df_clean)
STATS['adr_cap']           = adr_cap
STATS['cancel_rate_clean'] = df_clean['is_canceled'].mean() * 100

print(f'\n✅ Clean dataset : {len(df_clean):,} rows × {df_clean.shape[1]} columns')
print(f'   Removed       : {initial_rows - len(df_clean):,} rows  '
      f'({(initial_rows - len(df_clean)) / initial_rows * 100:.1f}% of raw)')
print(f'   Remaining NaN : {df_clean.isnull().sum().sum()}')
print(f'   Cancel rate   : {STATS["cancel_rate_clean"]:.1f}%  (was 37.0% in raw)')
print(f'   ADR capped at : ${adr_cap:.2f}  ({n_capped} rows affected)')
```

## 4. Exploratory Data Analysis (EDA)

> All EDA is performed on the **cleaned** dataset (`df_clean`).  
> `required_car_parking_spaces` and `assigned_room_type` appear here for documentation  
> but are **excluded from all ML models** (see Section 5).

---

```python
# ── 4.1  Cancellation rate by hotel type ──────────────────────────────────────
hotel_g = df_clean.groupby('hotel')['is_canceled'].agg(['sum', 'count'])
hotel_g['cancel_rate'] = (hotel_g['sum'] / hotel_g['count'] * 100).round(1)
hotel_g = hotel_g.reset_index()

city_rate   = hotel_g.loc[hotel_g['hotel'] == 'City Hotel',   'cancel_rate'].values[0]
resort_rate = hotel_g.loc[hotel_g['hotel'] == 'Resort Hotel', 'cancel_rate'].values[0]
STATS['city_cancel_rate']   = city_rate
STATS['resort_cancel_rate'] = resort_rate

print(hotel_g.to_string(index=False))
print(f'\n💡 Insight: City Hotels cancel at {city_rate}% vs Resort Hotels at {resort_rate}%.')
print('   City Hotels serve more transient business travellers who book flexible rates,')
print('   making them more prone to last-minute plan changes.')

fig = px.bar(
    hotel_g, x='hotel', y='cancel_rate', text='cancel_rate', color='hotel',
    title='Cancellation Rate by Hotel Type (%)',
    labels={'cancel_rate': 'Cancellation Rate (%)', 'hotel': 'Hotel Type'},
    color_discrete_sequence=['#3498db', '#e74c3c']
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

```python
# ── 4.2  Lead time vs cancellation ────────────────────────────────────────────
lead_g = df_clean.groupby('is_canceled')['lead_time'].mean().round(1)
lt_not = lead_g[0]
lt_yes = lead_g[1]
lt_diff = round((lt_yes - lt_not) / lt_not * 100, 1)
STATS['lt_not_canceled'] = lt_not
STATS['lt_canceled']     = lt_yes
STATS['lt_pct_longer']   = lt_diff

print(f'Average lead time — Not Canceled: {lt_not} days | Canceled: {lt_yes} days')
print(f'💡 Insight: Cancelled bookings have {lt_diff}% longer lead time.')
print('   The further ahead someone books, the more likely their plans change.')

fig = px.box(
    df_clean, x='is_canceled', y='lead_time', color='is_canceled',
    title='Lead Time Distribution: Canceled vs Not Canceled',
    labels={'is_canceled': 'Canceled (1=Yes)', 'lead_time': 'Lead Time (days)'},
    color_discrete_map={0: '#2ecc71', 1: '#e74c3c'}
)
fig.show()
```

```python
# ── 4.3  Deposit type ──────────────────────────────────────────────────────────
dep_g = df_clean.groupby('deposit_type')['is_canceled'].agg(['sum', 'count'])
dep_g['cancel_rate'] = (dep_g['sum'] / dep_g['count'] * 100).round(1)
dep_g = dep_g.reset_index()

nr_rate = dep_g.loc[dep_g['deposit_type'] == 'Non Refund', 'cancel_rate'].values[0]
STATS['non_refund_cancel_rate'] = nr_rate

print(dep_g[['deposit_type', 'count', 'sum', 'cancel_rate']].to_string(index=False))
print(f'\n💡 Non Refund = {nr_rate}% cancel rate.')
print('⚠️  Important nuance: Non-refundable bookings are often recorded as "Canceled"')
print('   in the hotel system even when the guest forfeits the deposit and the hotel')
print('   RETAINS the revenue. This is a data recording artefact, not a pure business')
print('   cancellation. The feature is kept but its interpretation requires caution.')

fig = px.bar(
    dep_g, x='deposit_type', y='cancel_rate', text='cancel_rate', color='deposit_type',
    title='Cancellation Rate by Deposit Type (%) — Note: Non Refund is a recording artefact',
    labels={'cancel_rate': 'Cancellation Rate (%)', 'deposit_type': 'Deposit Type'}
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

```python
# ── 4.4  Market segment ────────────────────────────────────────────────────────
mkt_g = df_clean.groupby('market_segment')['is_canceled'].agg(['sum', 'count'])
mkt_g['cancel_rate'] = (mkt_g['sum'] / mkt_g['count'] * 100).round(1)
mkt_g = mkt_g.sort_values('cancel_rate', ascending=False).reset_index()

real_mkt = mkt_g[mkt_g['market_segment'] != 'Undefined']
top_seg  = real_mkt.iloc[0]
safe_seg = real_mkt.iloc[-1]
STATS['top_market']    = f"{top_seg['market_segment']} ({top_seg['cancel_rate']}%)"
STATS['safest_market'] = f"{safe_seg['market_segment']} ({safe_seg['cancel_rate']}%)"

print(mkt_g.to_string(index=False))
print(f"\n💡 Highest-risk: {top_seg['market_segment']} ({top_seg['cancel_rate']}%)")
print(f"   Safest       : {safe_seg['market_segment']} ({safe_seg['cancel_rate']}%)")
print("   'Undefined' at 100% has only 2 bookings — a data label issue, not actionable.")

fig = px.bar(
    mkt_g, x='market_segment', y='cancel_rate', text='cancel_rate',
    color='cancel_rate',
    title='Cancellation Rate by Market Segment (%)',
    labels={'cancel_rate': 'Cancellation Rate (%)', 'market_segment': 'Market Segment'},
    color_continuous_scale='RdYlGn_r'
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

```python
# ── 4.5  Customer type ─────────────────────────────────────────────────────────
cust_g = df_clean.groupby('customer_type')['is_canceled'].agg(['sum', 'count'])
cust_g['cancel_rate'] = (cust_g['sum'] / cust_g['count'] * 100).round(1)
cust_g = cust_g.sort_values('cancel_rate', ascending=False).reset_index()

STATS['top_customer_type']    = f"{cust_g.iloc[0]['customer_type']} ({cust_g.iloc[0]['cancel_rate']}%)"
STATS['lowest_customer_type'] = f"{cust_g.iloc[-1]['customer_type']} ({cust_g.iloc[-1]['cancel_rate']}%)"

print(cust_g.to_string(index=False))
print(f"\n💡 Transient customers cancel most frequently — they book on flexible rates")
print(f"   through OTAs with easy cancellation policies.")

fig = px.bar(
    cust_g, x='customer_type', y='cancel_rate', text='cancel_rate', color='customer_type',
    title='Cancellation Rate by Customer Type (%)',
    labels={'cancel_rate': 'Cancellation Rate (%)', 'customer_type': 'Customer Type'}
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

```python
# ── 4.6  Monthly cancellation trend ───────────────────────────────────────────
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

mon_g = df_clean.groupby('arrival_date_month')['is_canceled'].agg(['sum', 'count'])
mon_g['cancel_rate'] = (mon_g['sum'] / mon_g['count'] * 100).round(1)
mon_g = mon_g.reindex(month_order).reset_index()

peak_idx  = mon_g['cancel_rate'].idxmax()
low_idx   = mon_g['cancel_rate'].idxmin()
peak_mon  = mon_g.loc[peak_idx, 'arrival_date_month']
peak_rate = mon_g.loc[peak_idx, 'cancel_rate']
low_mon   = mon_g.loc[low_idx,  'arrival_date_month']
low_rate  = mon_g.loc[low_idx,  'cancel_rate']
STATS['peak_month'] = peak_mon
STATS['peak_rate']  = peak_rate
STATS['low_month']  = low_mon
STATS['low_rate']   = low_rate

print(f'Peak cancellation  : {peak_mon} ({peak_rate}%)')
print(f'Lowest cancellation: {low_mon} ({low_rate}%)')
print(f'\n💡 Cancellation risk is highest in summer (high demand = guests shop around)')
print(f'   and lowest in late autumn/winter (more committed, niche bookings).')

fig = px.line(
    mon_g, x='arrival_date_month', y='cancel_rate', markers=True,
    title='Monthly Cancellation Rate (%)',
    labels={'arrival_date_month': 'Month', 'cancel_rate': 'Cancellation Rate (%)'}
)
fig.update_traces(line_color='#e74c3c', marker_size=8)
fig.show()
```

```python
# ── 4.7  Year-over-year cancellation trend (NEW) ──────────────────────────────
yr_g = df_clean.groupby('arrival_date_year')['is_canceled'].agg(['sum', 'count'])
yr_g['cancel_rate'] = (yr_g['sum'] / yr_g['count'] * 100).round(1)
yr_g = yr_g.reset_index()

print('Cancellation rate by year:')
print(yr_g[['arrival_date_year', 'count', 'cancel_rate']].to_string(index=False))
print('\n💡 Note: 2015 only covers July–Dec; 2017 only covers Jan–Aug.')
print('   Full-year comparisons should use 2016 as the reference year.')

fig = px.bar(
    yr_g, x='arrival_date_year', y='cancel_rate', text='cancel_rate',
    color='cancel_rate', color_continuous_scale='RdYlGn_r',
    title='Cancellation Rate by Arrival Year (%)',
    labels={'arrival_date_year': 'Year', 'cancel_rate': 'Cancellation Rate (%)'}
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

```python
# ── 4.8  ADR distribution vs cancellation ──────────────────────────────────────
adr_g = df_clean.groupby('is_canceled')['adr'].mean().round(2)
STATS['adr_not_canceled'] = adr_g[0]
STATS['adr_canceled']     = adr_g[1]

print(f'Average ADR — Not Canceled: ${adr_g[0]:.2f} | Canceled: ${adr_g[1]:.2f}')
print(f'💡 Cancelled bookings have ${adr_g[1] - adr_g[0]:.2f} higher ADR on average.')
print('   Premium-room guests may be shopping multiple properties simultaneously.')
print(f'   Reminder: ADR is capped at ${STATS["adr_cap"]:.2f} (99th pct) — extreme outliers removed.')

fig = px.histogram(
    df_clean, x='adr', color='is_canceled', barmode='overlay', nbins=60,
    title='ADR Distribution by Cancellation Status (capped at 99th percentile)',
    labels={'adr': 'Average Daily Rate ($)', 'is_canceled': 'Canceled'},
    color_discrete_map={0: '#2ecc71', 1: '#e74c3c'}, opacity=0.7
)
fig.show()
```

```python
# ── 4.9  Previous cancellations ────────────────────────────────────────────────
prev_g = df_clean.groupby('previous_cancellations')['is_canceled'].agg(['mean', 'count'])
prev_g['cancel_rate'] = (prev_g['mean'] * 100).round(1)
prev_g = prev_g[prev_g.index <= 5].reset_index()
prev_g.columns = ['previous_cancellations', 'mean', 'n_bookings', 'cancel_rate']

spike = prev_g.loc[prev_g['previous_cancellations'] == 1, 'cancel_rate'].values[0]
base  = prev_g.loc[prev_g['previous_cancellations'] == 0, 'cancel_rate'].values[0]
STATS['prev_cancel_spike'] = spike
STATS['prev_cancel_base']  = base

print(prev_g[['previous_cancellations', 'n_bookings', 'cancel_rate']].to_string(index=False))
print(f'\n💡 Guests with 1 prior cancellation re-cancel at {spike}% vs {base}% baseline.')
print('   This is the single strongest behavioural signal — flag these guests at booking.')
print('   Note: 2+ cancellations group is tiny (<200 bookings) — interpret cautiously.')

fig = px.bar(
    prev_g, x='previous_cancellations', y='cancel_rate',
    text=prev_g['cancel_rate'].round(1),
    title='Cancellation Rate by Number of Previous Cancellations',
    labels={'previous_cancellations': '# Previous Cancellations', 'cancel_rate': 'Cancellation Rate (%)'},
    color='cancel_rate', color_continuous_scale='Reds'
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

```python
# ── 4.10  Special requests vs cancellation ─────────────────────────────────────
spec_g = df_clean.groupby('total_of_special_requests')['is_canceled'].agg(['mean', 'count'])
spec_g['cancel_rate'] = (spec_g['mean'] * 100).round(1)
spec_g = spec_g.reset_index()
spec_g.columns = ['total_of_special_requests', 'mean', 'n_bookings', 'cancel_rate']

STATS['zero_req_cancel'] = spec_g.loc[spec_g['total_of_special_requests'] == 0, 'cancel_rate'].values[0]
STATS['high_req_cancel'] = spec_g['cancel_rate'].min()

print(spec_g[['total_of_special_requests', 'n_bookings', 'cancel_rate']].to_string(index=False))
print(f'\n💡 Zero special requests → {STATS["zero_req_cancel"]}% cancel rate.')
print(f'   High special requests  → as low as {STATS["high_req_cancel"]}% cancel rate.')
print('   More engagement at booking = greater commitment = lower cancellation.')
print('   Actionable: Prompt guests to add meal/room preferences to boost commitment.')

fig = px.bar(
    spec_g, x='total_of_special_requests', y='cancel_rate',
    text=spec_g['cancel_rate'].round(1),
    title='Cancellation Rate by Number of Special Requests',
    labels={'total_of_special_requests': '# Special Requests', 'cancel_rate': 'Cancellation Rate (%)'},
    color='cancel_rate', color_continuous_scale='RdYlGn_r'
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

## 5. Data Leak & Feature Audit

Before training any model, every candidate feature must pass a **temporal validity test**:  
*"Is this information available at the moment of prediction — i.e., when the booking is first created?"*

Features derived from post-booking events are **data leaks** and will produce inflated, non-generalisable metrics.

---

```python
print('=' * 72)
print('DATA LEAK AUDIT — ALL CANDIDATE FEATURES')
print('=' * 72)

audit = [
    ('reservation_status',        '🔴 DROPPED',           'Directly encodes the cancellation outcome — perfect leak'),
    ('reservation_status_date',   '🔴 DROPPED',           'Date of the outcome event — perfect leak'),
    ('assigned_room_type',        '🔴 EXCLUDED from model','Assigned at CHECK-IN not at booking (see 5.1)'),
    ('room_type_changed',         '🔴 EXCLUDED from model','Derived from assigned_room_type — inherits the leak'),
    ('required_car_parking_spaces','🔴 EXCLUDED from model','Check-in assignment artifact — 0% cancel rate (see 5.2)'),
    ('deposit_type',              '✅ INCLUDED',           'Known at booking; recording artefact documented'),
    ('lead_time',                 '✅ INCLUDED',           'Computed at booking creation time'),
    ('adr',                       '✅ INCLUDED (capped)',  'Rate agreed at booking; extreme outliers capped'),
    ('booking_changes',           '✅ INCLUDED*',          'Dynamic feature — 0 at creation; valid for ongoing monitoring'),
    ('market_segment',            '✅ INCLUDED',           'Known at booking creation time'),
    ('total_of_special_requests', '✅ INCLUDED',           'Known at booking creation time'),
    ('previous_cancellations',    '✅ INCLUDED',           'Historical record — known at booking time'),
    ('agent',                     '✅ INCLUDED',           'Known at booking creation time'),
    ('customer_type',             '✅ INCLUDED',           'Known at booking creation time'),
    ('meal',                      '✅ INCLUDED',           'Selected at booking time'),
    ('distribution_channel',      '✅ INCLUDED',           'Known at booking creation time'),
    ('hotel',                     '✅ INCLUDED',           'Known at booking creation time'),
    ('is_repeated_guest',         '✅ INCLUDED',           'Historical flag — known at booking time'),
    ('country',                   '✅ INCLUDED (grouped)', 'Known at booking; grouped to top-15 + Other'),
    ('arrival_date_month',        '✅ INCLUDED',           'Known at booking creation time — seasonality signal'),
]

print(f'  {"Feature":<35} {"Status":<25} Reason')
print('-' * 100)
for feat, status, reason in audit:
    print(f'  {feat:<35} {status:<25} {reason}')
print()
print('* booking_changes = 0 at booking creation; updates as booking matures.')
print('  Valid for real-time risk scoring throughout the booking lifecycle, but')
print('  interpretation at initial booking time is always 0 for new bookings.')
```

```python
# ── 5.1  Room type change — confirmed data leak ────────────────────────────────
print('INVESTIGATION 5.1 — Is assigned_room_type / room_type_changed a data leak?')
print('=' * 65)
print()
print('Logic:')
print('  • assigned_room_type is determined at CHECK-IN (not at booking time).')
print('  • A guest who CANCELS never reaches check-in.')
print('  • Therefore: if the assigned room differs from the reserved room,')
print('    that can ONLY happen for guests who actually checked in (= NOT canceled).')
print('  • room_type_changed=1 is a near-perfect proxy for "not canceled".')
print('  • Including it would make the model cheat — it could not generalise.')
print()

df_clean['_rtc'] = (df_clean['reserved_room_type'] != df_clean['assigned_room_type']).astype(int)
rtc_g = df_clean.groupby('_rtc')['is_canceled'].agg(['mean', 'count'])
rtc_g['cancel_rate'] = (rtc_g['mean'] * 100).round(2)
rtc_g = rtc_g.reset_index()
rtc_g.columns = ['room_type_changed', 'mean', 'n_bookings', 'cancel_rate']

print('Cancel rate by room_type_changed:')
print(rtc_g[['room_type_changed', 'n_bookings', 'cancel_rate']].to_string(index=False))

rate_chg   = rtc_g.loc[rtc_g['room_type_changed'] == 1, 'cancel_rate'].values[0]
rate_unchg = rtc_g.loc[rtc_g['room_type_changed'] == 0, 'cancel_rate'].values[0]
print(f'\n  Room NOT changed → {rate_unchg:.1f}% cancel rate')
print(f'  Room IS changed  → {rate_chg:.1f}% cancel rate')

if rate_chg < 5.0:
    print(f'\n🔴 CONFIRMED DATA LEAK: {rate_chg:.1f}% cancellation when room was changed.')
    print('   If the room was changed, that implies check-in occurred → not canceled.')
    print('   → assigned_room_type and room_type_changed EXCLUDED from all model features.')
else:
    print(f'\n⚠️  Unexpected: rate_chg={rate_chg:.1f}%. Investigate further.')

df_clean.drop(columns=['_rtc'], inplace=True)
```

```python
# ── 5.2  Parking spaces — confirmed check-in assignment artifact ───────────────
print('INVESTIGATION 5.2 — Why does parking≥1 show ~0% cancellation?')
print('=' * 65)
print()
print('CORRECTED explanation (v4 incorrectly attributed this to deduplication):')
print()
print('The raw dataset (119,390 rows — BEFORE any deduplication) ALSO shows')
print('~0% cancellation for bookings with parking≥1. Since the pattern exists')
print('in the raw data, deduplication cannot be the cause.')
print()
print('Most likely explanation (consistent with assigned_room_type findings):')
print('  • required_car_parking_spaces is ASSIGNED at CHECK-IN, not at booking.')
print('  • Guests who cancel never reach check-in.')
print('  • Their parking field therefore stays at 0 (never assigned).')
print('  • This is the SAME root cause as assigned_room_type — a check-in artifact.')
print()

# Demonstrate the pattern exists in RAW data too
raw_park = df.groupby('required_car_parking_spaces')['is_canceled'].agg(['mean', 'count'])
raw_park['cancel_rate'] = (raw_park['mean'] * 100).round(1)
raw_park = raw_park[raw_park.index <= 3].reset_index()
raw_park.columns = ['parking_spaces', 'mean', 'n_bookings', 'cancel_rate']
print('RAW data (no dedup) cancel rates by parking:')
print(raw_park[['parking_spaces', 'n_bookings', 'cancel_rate']].to_string(index=False))

clean_park = df_clean.groupby('required_car_parking_spaces')['is_canceled'].agg(['mean', 'count'])
clean_park['cancel_rate'] = (clean_park['mean'] * 100).round(1)
clean_park = clean_park[clean_park.index <= 3].reset_index()
clean_park.columns = ['parking_spaces', 'mean', 'n_bookings', 'cancel_rate']
print('\nCLEAN data (after dedup) cancel rates by parking:')
print(clean_park[['parking_spaces', 'n_bookings', 'cancel_rate']].to_string(index=False))

print()
print('🔴 Pattern is identical in both raw and clean data.')
print('   This is NOT a deduplication artifact — it is a check-in assignment artifact.')
print('   DECISION: required_car_parking_spaces EXCLUDED from the model.')
```

## 6. Feature Engineering & Encoding

**Design principles:**
- All features must be available at booking-creation time.
- Categorical encoders are **fitted once on the full dataset** and **saved** to disk.
- The same saved encoders are used in the Streamlit app — guaranteeing identical encoding.
- `required_car_parking_spaces`, `assigned_room_type`, and `room_type_changed` are excluded.

---

```python
df_ml = df_clean.copy()

# ── Derived features (all valid at booking creation time) ──────────────────────
df_ml['total_nights']    = df_ml['stays_in_weekend_nights'] + df_ml['stays_in_week_nights']
df_ml['total_guests']    = df_ml['adults'] + df_ml['children'] + df_ml['babies']
df_ml['has_agent']       = (df_ml['agent'] > 0).astype(int)
df_ml['is_weekend_only'] = (
    (df_ml['stays_in_weekend_nights'] > 0) & (df_ml['stays_in_week_nights'] == 0)
).astype(int)

# ── Country group — top-15 by volume + 'Other' (NEW) ──────────────────────────
TOP_N_COUNTRIES = 15
top_countries = df_ml['country'].value_counts().head(TOP_N_COUNTRIES).index.tolist()
df_ml['country_group'] = df_ml['country'].apply(
    lambda x: x if x in top_countries else 'Other'
)

print('Derived feature distributions:')
print(f'  total_nights     range: {df_ml["total_nights"].min()}–{df_ml["total_nights"].max()}'
      f'  (mean {df_ml["total_nights"].mean():.1f})')
print(f'  total_guests     range: {df_ml["total_guests"].min()}–{df_ml["total_guests"].max()}'
      f'  (mean {df_ml["total_guests"].mean():.1f})')
print(f'  has_agent        : {df_ml["has_agent"].value_counts().to_dict()}')
print(f'  is_weekend_only  : {df_ml["is_weekend_only"].value_counts().to_dict()}')
print(f'  country_group top values: {df_ml["country_group"].value_counts().head(5).to_dict()}')
```

```python
# ── Encoding — fit once, save encoders for Streamlit ──────────────────────────
fitted_encoders = {}    # will be saved to disk for the Streamlit app

# 1. Lead-time bucket — ORDINAL (preserves order)
lead_bins   = [-1, 7, 30, 90, 180, 365, 9999]
lead_labels = ['0-7d', '8-30d', '31-90d', '91-180d', '181-365d', '365+d']
df_ml['lead_time_bucket_cat'] = pd.cut(df_ml['lead_time'], bins=lead_bins, labels=lead_labels)

LT_ORDER = [['0-7d', '8-30d', '31-90d', '91-180d', '181-365d', '365+d']]
oe_lead = OrdinalEncoder(categories=LT_ORDER, handle_unknown='use_encoded_value', unknown_value=-1)
df_ml['lead_time_bucket'] = oe_lead.fit_transform(
    df_ml[['lead_time_bucket_cat']].astype(str)
).flatten().astype(int)
fitted_encoders['lead_time_bucket'] = oe_lead

# 2. Arrival month — ORDINAL (preserves seasonal order)
MONTH_ORDER = [['January','February','March','April','May','June',
                'July','August','September','October','November','December']]
oe_month = OrdinalEncoder(categories=MONTH_ORDER, handle_unknown='use_encoded_value', unknown_value=-1)
df_ml['arrival_month_enc'] = oe_month.fit_transform(
    df_ml[['arrival_date_month']]
).flatten().astype(int)
fitted_encoders['arrival_date_month'] = oe_month

# 3. Nominal categoricals — LABEL ENCODER
NOM_CATS = ['hotel', 'meal', 'market_segment', 'distribution_channel',
            'deposit_type', 'customer_type', 'country_group']
for col in NOM_CATS:
    le = LabelEncoder()
    df_ml[col] = le.fit_transform(df_ml[col].astype(str))
    fitted_encoders[col] = le
    print(f'  {col:<25}: {len(le.classes_)} unique values → encoded')

print('\n✅ All encoders fitted.')
print('   Encoders will be saved in Section 12 for consistent Streamlit app inference.')
```

```python
# ── Define feature sets ────────────────────────────────────────────────────────
# FULL — includes deposit_type (recommended for deployment)
FEATURES_FULL = [
    'lead_time', 'total_nights', 'total_guests', 'adr',
    'previous_cancellations', 'previous_bookings_not_canceled',
    'booking_changes', 'days_in_waiting_list',
    'total_of_special_requests', 'is_repeated_guest',
    'has_agent', 'is_weekend_only',
    'hotel', 'meal', 'market_segment', 'distribution_channel',
    'deposit_type', 'customer_type',
    'lead_time_bucket', 'arrival_month_enc', 'country_group',  # 3 features vs v4
]

# NO-DEPOSIT — for sensitivity analysis
FEATURES_NO_DEP = [f for f in FEATURES_FULL if f != 'deposit_type']

TARGET = 'is_canceled'

X_full  = df_ml[FEATURES_FULL]
X_nodep = df_ml[FEATURES_NO_DEP]
y       = df_ml[TARGET]

STATS['n_features'] = len(FEATURES_FULL)

print(f'Feature count (full set)    : {len(FEATURES_FULL)}')
print(f'Feature count (no deposit)  : {len(FEATURES_NO_DEP)}')
print(f'All features: {FEATURES_FULL}')
```

```python
# ── Train / Test split  (80 / 20, stratified) ─────────────────────────────────
X_train_f, X_test_f, y_train, y_test = train_test_split(
    X_full, y, test_size=0.20, random_state=42, stratify=y
)
X_train_nd, X_test_nd, _, _ = train_test_split(
    X_nodep, y, test_size=0.20, random_state=42, stratify=y
)

STATS['train_rows'] = X_train_f.shape[0]
STATS['test_rows']  = X_test_f.shape[0]

print(f'Training rows : {X_train_f.shape[0]:,}')
print(f'Test rows     : {X_test_f.shape[0]:,}')
print()
test_bal = y_test.value_counts(normalize=True)
print(f'Test class balance — Not Canceled: {test_bal[0]:.1%}  Canceled: {test_bal[1]:.1%}')
print()
print('💡 Class imbalance: ~72.5% not-canceled vs ~27.5% canceled after deduplication.')
print('   Imbalance is handled via:')
print('   • class_weight="balanced" for Logistic Regression')
print('   • scale_pos_weight for XGBoost')
print('   • Threshold tuning in Section 8')
print('   NOT by oversampling — that would artificially inflate metrics.')
```

## 7. Model Training & Evaluation

```python
def evaluate_model(name, model, X_tr, X_te, y_tr, y_te):
    """Fit model and return metric dict, fitted model, predictions, probabilities."""
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_te)
    y_prob = model.predict_proba(X_te)[:, 1]
    metrics = {
        'Model'    : name,
        'Accuracy' : accuracy_score(y_te, y_pred),
        'Precision': precision_score(y_te, y_pred, zero_division=0),
        'Recall'   : recall_score(y_te, y_pred, zero_division=0),
        'F1 Score' : f1_score(y_te, y_pred, zero_division=0),
        'ROC-AUC'  : roc_auc_score(y_te, y_prob),
    }
    return metrics, model, y_pred, y_prob

all_results    = []
trained_models = {}   # name → (model, y_pred, y_prob)
```

```python
# ── Logistic Regression ────────────────────────────────────────────────────────
# Requires scaled features; class_weight='balanced' handles class imbalance
scaler     = StandardScaler()
X_train_sc = scaler.fit_transform(X_train_f)
X_test_sc  = scaler.transform(X_test_f)

m, mdl, yp, ypr = evaluate_model(
    'Logistic Regression',
    LogisticRegression(max_iter=2000, class_weight='balanced', C=0.5, random_state=42),
    X_train_sc, X_test_sc, y_train, y_test
)
all_results.append(m)
trained_models['Logistic Regression'] = (mdl, yp, ypr)
print(f"Logistic Regression — Accuracy: {m['Accuracy']:.4f}  ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
# ── Decision Tree ──────────────────────────────────────────────────────────────
m, mdl, yp, ypr = evaluate_model(
    'Decision Tree',
    DecisionTreeClassifier(max_depth=8, min_samples_leaf=50,
                           class_weight='balanced', random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m)
trained_models['Decision Tree'] = (mdl, yp, ypr)
print(f"Decision Tree       — Accuracy: {m['Accuracy']:.4f}  ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
# ── Random Forest ──────────────────────────────────────────────────────────────
m, mdl, yp, ypr = evaluate_model(
    'Random Forest',
    RandomForestClassifier(n_estimators=200, max_depth=12, min_samples_leaf=20,
                           class_weight='balanced', n_jobs=-1, random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m)
trained_models['Random Forest'] = (mdl, yp, ypr)
print(f"Random Forest       — Accuracy: {m['Accuracy']:.4f}  ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
# ── Gradient Boosting ──────────────────────────────────────────────────────────
m, mdl_gb, yp_gb, ypr_gb = evaluate_model(
    'Gradient Boosting',
    GradientBoostingClassifier(n_estimators=300, learning_rate=0.08,
                               max_depth=5, subsample=0.8, random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m)
trained_models['Gradient Boosting'] = (mdl_gb, yp_gb, ypr_gb)
m_gb = m
print(f"Gradient Boosting   — Accuracy: {m['Accuracy']:.4f}  ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
# ── XGBoost ────────────────────────────────────────────────────────────────────
if XGBOOST_AVAILABLE:
    scale_pw = float((y_train == 0).sum()) / float((y_train == 1).sum())
    m, mdl_xgb, yp_xgb, ypr_xgb = evaluate_model(
        'XGBoost',
        XGBClassifier(
            n_estimators=300, learning_rate=0.08, max_depth=6,
            subsample=0.8, colsample_bytree=0.8,
            scale_pos_weight=scale_pw,
            eval_metric='logloss', n_jobs=-1, random_state=42,
            verbosity=0
        ),
        X_train_f, X_test_f, y_train, y_test
    )
    all_results.append(m)
    trained_models['XGBoost'] = (mdl_xgb, yp_xgb, ypr_xgb)
    print(f"XGBoost             — Accuracy: {m['Accuracy']:.4f}  ROC-AUC: {m['ROC-AUC']:.4f}")
else:
    print('XGBoost skipped — run the install cell at the top and restart kernel.')
```

```python
# ── Results comparison table ───────────────────────────────────────────────────
results_df = pd.DataFrame(all_results).set_index('Model').sort_values('ROC-AUC', ascending=False)

STATS['best_model']    = results_df.index[0]
STATS['best_roc_auc']  = results_df['ROC-AUC'].iloc[0]
STATS['best_accuracy'] = results_df['Accuracy'].iloc[0]
STATS['best_recall']   = results_df['Recall'].iloc[0]
STATS['best_f1']       = results_df['F1 Score'].iloc[0]

print('Model Comparison — v5 (21 clean features, includes country + month):')
print(results_df.round(4).to_string())
print()
print(f'🏆 Best model: {STATS["best_model"]}  ROC-AUC: {STATS["best_roc_auc"]:.4f}')
print()
print('📝 Performance context:')
print('  • v3 ROC-AUC: 0.8605  (inflated — had data-leaky features)')
print('  • v4 ROC-AUC: 0.8257  (honest — removed leaky features)')
print(f'  • v5 ROC-AUC: {STATS["best_roc_auc"]:.4f}  (honest — added country + month features)')
print('  A higher but DISHONEST score is worthless. These numbers are trustworthy.')
```

## 8. Threshold Tuning — Balancing Recall and Precision

The default decision threshold of 0.5 is conservative: it maximises accuracy on balanced data  
but **misses many cancellations** when the class distribution is imbalanced (~72/28 here).  

Lowering the threshold catches more cancellations (higher Recall) at the cost of more false alarms (lower Precision).  
**Hotels choose their threshold based on their overbooking tolerance.**

> ⚠️ Note: The threshold sweep below is shown on the test set for transparency.  
> In production, select the threshold using cross-validation on training data to avoid  
> any information leakage from the test set into the decision process.

---

```python
best_model_name = STATS['best_model']
best_prob       = trained_models[best_model_name][2]  # probabilities on test set

print(f'Threshold sweep for: {best_model_name}')
print()

thresholds = np.arange(0.15, 0.80, 0.05)
thr_rows   = []

for thr in thresholds:
    yp_thr = (best_prob >= thr).astype(int)
    thr_rows.append({
        'Threshold' : round(float(thr), 2),
        'Accuracy'  : accuracy_score(y_test, yp_thr),
        'Precision' : precision_score(y_test, yp_thr, zero_division=0),
        'Recall'    : recall_score(y_test, yp_thr, zero_division=0),
        'F1 Score'  : f1_score(y_test, yp_thr, zero_division=0),
    })

thr_df = pd.DataFrame(thr_rows)
print(thr_df.round(3).to_string(index=False))

best_f1_thr = thr_df.loc[thr_df['F1 Score'].idxmax(), 'Threshold']
STATS['recommended_threshold'] = best_f1_thr
print(f'\n📌 Threshold with best F1 Score        : {best_f1_thr}')
print(  '   Threshold for high-recall (hotel-safe): 0.25–0.35')
print(  '   Threshold for high-precision           : 0.50+')
```

```python
# ── Threshold trade-off chart ──────────────────────────────────────────────────
fig = go.Figure()
for metric, color in [('Recall',    '#e74c3c'), ('Precision', '#3498db'),
                       ('F1 Score', '#2ecc71'),  ('Accuracy',  '#f39c12')]:
    fig.add_trace(go.Scatter(
        x=thr_df['Threshold'], y=thr_df[metric],
        mode='lines+markers', name=metric, line=dict(color=color, width=2)
    ))

fig.add_vline(x=0.5, line_dash='dash', line_color='gray',
              annotation_text='Default (0.5)', annotation_position='top right')
fig.add_vline(x=best_f1_thr, line_dash='dot', line_color='red',
              annotation_text=f'Best F1 ({best_f1_thr})', annotation_position='top left')
fig.update_layout(
    title=f'Threshold Tuning — {best_model_name}',
    xaxis_title='Decision Threshold',
    yaxis_title='Score',
    height=450
)
fig.show()

# ── Final report at recommended threshold ─────────────────────────────────────
y_pred_rec = (best_prob >= best_f1_thr).astype(int)
print(f'\nClassification report at threshold = {best_f1_thr} (best F1 / hotel-recommended):')
print(classification_report(y_test, y_pred_rec, target_names=['Not Canceled', 'Canceled']))
```

```python
# ── 5-Fold Stratified Cross-Validation ────────────────────────────────────────
# Use a faster GB config for CV (n_estimators=150 vs 300) to keep runtime reasonable
print(f'Running 5-Fold CV for {best_model_name}...')
cv_model  = GradientBoostingClassifier(
    n_estimators=150, learning_rate=0.08, max_depth=5, subsample=0.8, random_state=42
)
skf       = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(cv_model, X_full, y, cv=skf, scoring='roc_auc', n_jobs=-1)

STATS['cv_mean'] = cv_scores.mean()
STATS['cv_std']  = cv_scores.std()

print(f'ROC-AUC per fold : {cv_scores.round(4)}')
print(f'Mean ROC-AUC     : {cv_scores.mean():.4f}')
print(f'Std Dev          : {cv_scores.std():.4f}')

if cv_scores.std() < 0.01:
    print(f'✅ Std dev {cv_scores.std():.4f} < 0.01 — model is very stable.')
elif cv_scores.std() < 0.02:
    print(f'✅ Std dev {cv_scores.std():.4f} — model is stable.')
else:
    print(f'⚠️  Std dev {cv_scores.std():.4f} — some variance; consider regularisation.')

gap = STATS['best_roc_auc'] - cv_scores.mean()
print(f'\nTest AUC vs CV mean: {STATS["best_roc_auc"]:.4f} vs {cv_scores.mean():.4f}  (gap={gap:.4f})')
# A gap up to 0.015 is acceptable — larger gaps suggest lucky/unlucky test split
if abs(gap) <= 0.015:
    print('✅ Gap ≤ 0.015 — test result is consistent with cross-validation.')
else:
    print('⚠️  Gap > 0.015 — review train/test split randomness.')

fig = px.bar(
    x=[f'Fold {i+1}' for i in range(5)], y=cv_scores,
    text=cv_scores.round(4),
    title=f'5-Fold CV ROC-AUC — {best_model_name} (v5)',
    labels={'x': 'Fold', 'y': 'ROC-AUC'},
    color=cv_scores, color_continuous_scale='Blues'
)
fig.add_hline(y=cv_scores.mean(), line_dash='dash', line_color='red',
              annotation_text=f'Mean={cv_scores.mean():.4f}')
fig.update_traces(texttemplate='%{text}', textposition='outside')
fig.show()
```

## 11. SHAP Value Analysis — Model Explainability

SHAP (SHapley Additive exPlanations) explains **why** the model made each prediction.  
This is essential for the Streamlit app (per-booking explanations) and for hotel staff trust.

Install: `pip install shap` (run the install cell at top and restart kernel if needed).

---

```python
if not SHAP_AVAILABLE:
    print('⚠️  SHAP not available.')
    print('   Run the install cell at the top of this notebook and restart the kernel.')
    print('   Then re-run this section.')
else:
    print('Computing SHAP values on a 2,000-row sample of the test set...')
    rng        = np.random.RandomState(42)
    sample_idx = rng.choice(len(X_test_f), size=min(2000, len(X_test_f)), replace=False)
    X_shap     = X_test_f.iloc[sample_idx].copy()

    explainer   = shap.TreeExplainer(mdl_gb)
    shap_values = explainer.shap_values(X_shap)

    print(f'✅ SHAP values computed. Shape: {shap_values.shape}')
    print()

    mean_shap = pd.DataFrame({
        'Feature'    : X_shap.columns,
        'Mean |SHAP|': np.abs(shap_values).mean(axis=0)
    }).sort_values('Mean |SHAP|', ascending=False)

    STATS['shap_top_feature'] = mean_shap.iloc[0]['Feature']
    print('Global importance via mean |SHAP| value:')
    print(mean_shap.to_string(index=False))
```

```python
if SHAP_AVAILABLE:
    print('SHAP Beeswarm Summary Plot')
    print('Red = high feature value, Blue = low feature value')
    print('Right of centre line = pushes prediction toward cancellation')
    plt.figure(figsize=(10, 8))
    shap.summary_plot(shap_values, X_shap, max_display=15, show=True)

    print('\nSHAP Bar Plot — Mean Absolute Feature Impact')
    plt.figure(figsize=(10, 6))
    shap.summary_plot(shap_values, X_shap, plot_type='bar', max_display=15, show=True)
```

```python
if SHAP_AVAILABLE:
    # Waterfall plot for a high-risk booking
    best_prob_sample = trained_models[best_model_name][2]
    best_prob_shap   = best_prob_sample[sample_idx]
    high_risk_mask   = best_prob_shap > 0.65

    if high_risk_mask.any():
        idx = np.where(high_risk_mask)[0][0]
        print(f'Waterfall plot for a high-risk booking')
        print(f'  Predicted cancel probability: {best_prob_shap[idx]:.3f}')
        print(f'  Actual outcome              : {"Canceled" if y_test.iloc[sample_idx[idx]]==1 else "Not Canceled"}')
        print()

        explanation = shap.Explanation(
            values       = shap_values[idx],
            base_values  = explainer.expected_value,
            data         = X_shap.iloc[idx].values,
            feature_names= list(X_shap.columns)
        )
        plt.figure(figsize=(10, 6))
        shap.waterfall_plot(explanation, max_display=12, show=True)
    else:
        print('No bookings with predicted probability > 0.65 in the sample.')
        print('Showing first booking in the sample instead.')
        explanation = shap.Explanation(
            values       = shap_values[0],
            base_values  = explainer.expected_value,
            data         = X_shap.iloc[0].values,
            feature_names= list(X_shap.columns)
        )
        plt.figure(figsize=(10, 6))
        shap.waterfall_plot(explanation, max_display=12, show=True)
```
