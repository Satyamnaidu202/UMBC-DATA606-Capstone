## 3. Data Cleaning & Preprocessing

```python
# Check missing values
missing = df.isnull().sum()
miss_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing / len(df) * 100})
miss_df = miss_df[miss_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)
print(miss_df.to_string())

fig = px.bar(
    miss_df.reset_index(), x='index', y='Missing %',
    title='Missing Value Percentage by Column',
    labels={'index': 'Column'},
    color='Missing %', color_continuous_scale='Reds'
)
fig.show()
```

```python
# Duplicate analysis — duplicates are disproportionately canceled bookings,
# which is why the overall cancellation rate drops after removal
dupes    = df.duplicated().sum()
dup_mask = df.duplicated(keep=False)

print(f'Exact duplicate rows   : {dupes:,}  ({dupes / len(df) * 100:.1f}% of raw data)')
print(f'Cancel rate — duplicated rows    : {df.loc[dup_mask,  "is_canceled"].mean()*100:.1f}%')
print(f'Cancel rate — non-duplicated rows: {df.loc[~dup_mask, "is_canceled"].mean()*100:.1f}%')
```

```python
df_clean     = df.copy()
initial_rows = len(df_clean)

# Step 1 — Remove exact duplicates (system re-entries, not distinct bookings)
df_clean.drop_duplicates(inplace=True)
print(f'Step 1 – Duplicates removed       : -{initial_rows - len(df_clean):,}  → {len(df_clean):,} rows')

# Step 2 — Remove bookings with zero guests (impossible in reality)
before = len(df_clean)
df_clean = df_clean[(df_clean['adults'] + df_clean['children'].fillna(0) + df_clean['babies']) > 0]
print(f'Step 2 – Zero-guest rows removed  : -{before - len(df_clean):,}  → {len(df_clean):,} rows')

# Step 3 — Remove negative ADR (impossible price)
before = len(df_clean)
df_clean = df_clean[df_clean['adr'] >= 0]
print(f'Step 3 – Negative ADR removed     : -{before - len(df_clean):,}  → {len(df_clean):,} rows')

# Step 4 — Drop columns that cannot be used
#   company              → 94% missing, not recoverable
#   reservation_status   → directly encodes the cancellation outcome (data leak)
#   reservation_status_date → the date of that outcome (data leak)
df_clean.drop(columns=['company', 'reservation_status', 'reservation_status_date'], inplace=True)
print('Step 4 – Dropped: company, reservation_status, reservation_status_date')

# Step 5 — Impute remaining missing values
df_clean['children'].fillna(0, inplace=True)        # 4 rows
df_clean['country'].fillna('Unknown', inplace=True)  # 488 rows
df_clean['agent'].fillna(0, inplace=True)            # 16,340 rows — 0 means no agent

# Step 6 — Cap ADR at the 99th percentile to remove extreme outliers
adr_cap  = df_clean['adr'].quantile(0.99)
n_capped = (df_clean['adr'] > adr_cap).sum()
df_clean['adr'] = df_clean['adr'].clip(upper=adr_cap)
print(f'Step 6 – ADR capped at ${adr_cap:.2f} (99th pct): {n_capped} rows affected')

STATS['clean_rows']        = len(df_clean)
STATS['removed_rows']      = initial_rows - len(df_clean)
STATS['adr_cap']           = adr_cap
STATS['cancel_rate_clean'] = df_clean['is_canceled'].mean() * 100

print(f'\nClean dataset : {len(df_clean):,} rows × {df_clean.shape[1]} columns')
print(f'Cancel rate   : {STATS["cancel_rate_clean"]:.1f}%  (was 37.0% in raw — reduced after deduplication)')
print(f'Remaining NaN : {df_clean.isnull().sum().sum()}')
```

<a id='5'></a>
## 5. Data Leak & Feature Audit

Before training any model, every feature must pass a temporal validity test:

> *"Is this information available at the moment the booking is first created?"*

Features that are only known after the booking outcome are **data leaks** — they produce inflated metrics that will not generalise to real-world predictions.

| Feature | Decision | Reason |
|---------|----------|--------|
| `reservation_status` | ❌ Dropped | Directly encodes the cancellation outcome |
| `reservation_status_date` | ❌ Dropped | Date of the outcome event |
| `assigned_room_type` | ❌ Excluded | Assigned at check-in, not at booking time |
| `room_type_changed` | ❌ Excluded | Derived from `assigned_room_type` — inherits the leak |
| `required_car_parking_spaces` | ❌ Excluded | Assigned at check-in — same root cause |
| `company` | ❌ Dropped | 94% missing — not recoverable |
| `deposit_type` | ✅ Included | Known at booking; recording artefact documented |
| `lead_time` | ✅ Included | Computed at booking creation time |
| `adr` | ✅ Included (capped) | Rate agreed at booking |
| `booking_changes` | ✅ Included* | 0 at creation; valid for ongoing monitoring |
| `market_segment` | ✅ Included | Known at booking |
| `total_of_special_requests` | ✅ Included | Known at booking |
| `previous_cancellations` | ✅ Included | Historical record known at booking |
| `agent` | ✅ Included | Known at booking |
| `customer_type` | ✅ Included | Known at booking |
| `meal` | ✅ Included | Selected at booking |
| `distribution_channel` | ✅ Included | Known at booking |
| `hotel` | ✅ Included | Known at booking |
| `is_repeated_guest` | ✅ Included | Historical flag |
| `country` | ✅ Included (grouped) | Known at booking |
| `arrival_date_month` | ✅ Included | Known at booking — captures seasonality |

*`booking_changes` is 0 for all new bookings; it updates as the booking matures and is valid for real-time risk monitoring.*

### 5.1 Room Type Change — Confirmed Data Leak

```python
# Create a temporary flag: 1 if the assigned room differs from the reserved room
df_clean['_rtc'] = (df_clean['reserved_room_type'] != df_clean['assigned_room_type']).astype(int)
rtc_g = df_clean.groupby('_rtc')['is_canceled'].agg(['mean', 'count'])
rtc_g['cancel_rate'] = (rtc_g['mean'] * 100).round(2)
rtc_g.index.name = 'room_type_changed'

rate_chg   = rtc_g.loc[1, 'cancel_rate']
rate_unchg = rtc_g.loc[0, 'cancel_rate']

print(rtc_g[['count', 'cancel_rate']].to_string())
print(f'\nRoom NOT changed → {rate_unchg:.1f}% cancel rate')
print(f'Room IS changed  → {rate_chg:.1f}% cancel rate')
print('\nLogic: A guest who cancels never reaches check-in, so their room can never be re-assigned.')
print('room_type_changed = 1 is therefore a near-perfect proxy for "not canceled" — a data leak.')
print('Decision: assigned_room_type and room_type_changed are excluded from all model features.')

df_clean.drop(columns=['_rtc'], inplace=True)
```

### 5.2 Car Parking — Confirmed Check-in Assignment Artifact

```python
# The ~0% cancellation rate for parking≥1 needs explanation.
# Key finding: the pattern exists in the raw data BEFORE deduplication,
# so deduplication cannot be the cause.
# Root cause: parking spaces are assigned at check-in, not at booking time.
# Guests who cancel never check in, so their parking field stays at 0.
# This is the same mechanism as assigned_room_type.

raw_rate   = df[df['required_car_parking_spaces'] >= 1]['is_canceled'].mean() * 100
clean_rate = df_clean[df_clean['required_car_parking_spaces'] >= 1]['is_canceled'].mean() * 100

print(f'Raw data  (before dedup) — bookings with parking≥1: cancel rate = {raw_rate:.1f}%')
print(f'Clean data (after dedup) — bookings with parking≥1: cancel rate = {clean_rate:.1f}%')
print('\nThe pattern is identical in both datasets — this is a check-in assignment artifact, not a deduplication artifact.')
print('Decision: required_car_parking_spaces is excluded from all models.')
```

<a id='6'></a>
## 6. Feature Engineering & Encoding

**Design principles:**
- All features must be available at booking-creation time
- Encoders are fitted once on the full dataset and saved to disk — the Streamlit app loads the same objects, guaranteeing identical encoding at inference time
- **21 final features:** 12 raw numeric/binary + 7 label-encoded categoricals + 2 ordinal-encoded fields (`lead_time_bucket`, `arrival_month_enc`)

```python
df_ml = df_clean.copy()

# Derived features — all computable at booking creation time
df_ml['total_nights']    = df_ml['stays_in_weekend_nights'] + df_ml['stays_in_week_nights']
df_ml['total_guests']    = df_ml['adults'] + df_ml['children'] + df_ml['babies']
df_ml['has_agent']       = (df_ml['agent'] > 0).astype(int)
df_ml['is_weekend_only'] = (
    (df_ml['stays_in_weekend_nights'] > 0) & (df_ml['stays_in_week_nights'] == 0)
).astype(int)

# Group countries: top-15 by booking volume kept individually; all others → 'Other'
# This avoids high-cardinality encoding while preserving signal from frequent countries
TOP_N_COUNTRIES = 15
top_countries  = df_ml['country'].value_counts().head(TOP_N_COUNTRIES).index.tolist()
df_ml['country_group'] = df_ml['country'].apply(lambda x: x if x in top_countries else 'Other')

print('Derived features summary:')
print(f'  total_nights  range: {df_ml["total_nights"].min()}–{df_ml["total_nights"].max()}  (mean {df_ml["total_nights"].mean():.1f})')
print(f'  total_guests  range: {df_ml["total_guests"].min()}–{df_ml["total_guests"].max()}  (mean {df_ml["total_guests"].mean():.1f})')
print(f'  has_agent     : {df_ml["has_agent"].value_counts().to_dict()}')
print(f'  is_weekend_only: {df_ml["is_weekend_only"].value_counts().to_dict()}')
print(f'  country_group top-5: {df_ml["country_group"].value_counts().head(5).to_dict()}')
```

```python
# All fitted encoders are stored here and saved to disk for the Streamlit app
fitted_encoders = {}

# 1. Lead-time bucket — ordinal encoding preserves the natural order of time ranges
lead_bins   = [-1, 7, 30, 90, 180, 365, 9999]
lead_labels = ['0-7d', '8-30d', '31-90d', '91-180d', '181-365d', '365+d']
df_ml['lead_time_bucket_cat'] = pd.cut(df_ml['lead_time'], bins=lead_bins, labels=lead_labels)

oe_lead = OrdinalEncoder(categories=[lead_labels], handle_unknown='use_encoded_value', unknown_value=-1)
df_ml['lead_time_bucket'] = oe_lead.fit_transform(
    df_ml[['lead_time_bucket_cat']].astype(str)
).flatten().astype(int)
fitted_encoders['lead_time_bucket'] = oe_lead

# 2. Arrival month — ordinal encoding preserves seasonal order (Jan=0 … Dec=11)
MONTH_ORDER = [['January','February','March','April','May','June',
                'July','August','September','October','November','December']]
oe_month = OrdinalEncoder(categories=MONTH_ORDER, handle_unknown='use_encoded_value', unknown_value=-1)
df_ml['arrival_month_enc'] = oe_month.fit_transform(
    df_ml[['arrival_date_month']]
).flatten().astype(int)
fitted_encoders['arrival_date_month'] = oe_month

# 3. Nominal categoricals — label encoding (no natural order)
NOM_CATS = ['hotel', 'meal', 'market_segment', 'distribution_channel',
            'deposit_type', 'customer_type', 'country_group']
for col in NOM_CATS:
    le = LabelEncoder()
    df_ml[col] = le.fit_transform(df_ml[col].astype(str))
    fitted_encoders[col] = le
    print(f'  {col:<25}: {len(le.classes_)} unique values')

print('\nAll encoders fitted and ready to save.')
```

```python
# Final feature list — order matters for model inference
FEATURES_FULL = [
    'lead_time', 'total_nights', 'total_guests', 'adr',
    'previous_cancellations', 'previous_bookings_not_canceled',
    'booking_changes', 'days_in_waiting_list',
    'total_of_special_requests', 'is_repeated_guest',
    'has_agent', 'is_weekend_only',
    'hotel', 'meal', 'market_segment', 'distribution_channel',
    'deposit_type', 'customer_type',
    'lead_time_bucket', 'arrival_month_enc', 'country_group',
]

# No-deposit set used for sensitivity analysis in Section 9
FEATURES_NO_DEP = [f for f in FEATURES_FULL if f != 'deposit_type']

TARGET = 'is_canceled'

X_full  = df_ml[FEATURES_FULL]
X_nodep = df_ml[FEATURES_NO_DEP]
y       = df_ml[TARGET]
STATS['n_features'] = len(FEATURES_FULL)

print(f'Full feature set : {len(FEATURES_FULL)} features')
print(f'No-deposit set   : {len(FEATURES_NO_DEP)} features')
print(f'\nAll features: {FEATURES_FULL}')
```

```python
# 80/20 stratified split — stratify=y ensures the class ratio is preserved in both sets
X_train_f, X_test_f, y_train, y_test = train_test_split(
    X_full, y, test_size=0.20, random_state=42, stratify=y)
X_train_nd, X_test_nd, _, _ = train_test_split(
    X_nodep, y, test_size=0.20, random_state=42, stratify=y)

STATS['train_rows'] = X_train_f.shape[0]
STATS['test_rows']  = X_test_f.shape[0]
test_bal = y_test.value_counts(normalize=True)

print(f'Training rows : {X_train_f.shape[0]:,}')
print(f'Test rows     : {X_test_f.shape[0]:,}')
print(f'Class balance (test) — Not Canceled: {test_bal[0]:.1%}  |  Canceled: {test_bal[1]:.1%}')
print('\nClass imbalance (~72.5% / 27.5%) is handled via:')
print('  class_weight="balanced" for Logistic Regression, Decision Tree, Random Forest')
print('  scale_pos_weight for XGBoost')
print('  Oversampling is intentionally avoided — it would artificially inflate metrics.')
```

<a id='7'></a>
## 7. Model Training & Evaluation

Five classifiers are trained on identical 80/20 stratified splits.

**Primary metric: ROC-AUC** — threshold-independent and robust to class imbalance.  
**Secondary focus: Recall and F1** — missed cancellations (False Negatives) directly cost revenue.

| Model | Class Imbalance Strategy |
|-------|-------------------------|
| Logistic Regression | `class_weight='balanced'` |
| Decision Tree | `class_weight='balanced'` |
| Random Forest | `class_weight='balanced'` |
| Gradient Boosting | Ensemble naturally handles mild imbalance |
| XGBoost | `scale_pos_weight ≈ 2.63` — achieves high recall at the default 0.5 threshold |

```python
def evaluate_model(name, model, X_tr, X_te, y_tr, y_te):
    """Fit model, predict on test set, and return metrics, fitted model, predictions, and probabilities."""
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_te)
    y_prob = model.predict_proba(X_te)[:, 1]
    metrics = {
        'Model'    : name,
        'Accuracy' : accuracy_score(y_te, y_pred),
        'Precision': precision_score(y_te, y_pred, zero_division=0),
        'Recall'   : recall_score(y_te, y_pred, zero_division=0),
        'F1 Score' : f1_score(y_te, y_pred, zero_division=0),
        'ROC-AUC'  : roc_auc_score(y_te, y_prob),
    }
    return metrics, model, y_pred, y_prob

all_results    = []
trained_models = {}
```

```python
# Logistic Regression requires feature scaling
scaler     = StandardScaler()
X_train_sc = scaler.fit_transform(X_train_f)
X_test_sc  = scaler.transform(X_test_f)

m, mdl, yp, ypr = evaluate_model(
    'Logistic Regression',
    LogisticRegression(max_iter=2000, class_weight='balanced', C=0.5, random_state=42),
    X_train_sc, X_test_sc, y_train, y_test
)
all_results.append(m)
trained_models['Logistic Regression'] = (mdl, yp, ypr)
print(f"Logistic Regression — Accuracy: {m['Accuracy']:.4f}  ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
m, mdl, yp, ypr = evaluate_model(
    'Decision Tree',
    DecisionTreeClassifier(max_depth=8, min_samples_leaf=50, class_weight='balanced', random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m)
trained_models['Decision Tree'] = (mdl, yp, ypr)
print(f"Decision Tree       — Accuracy: {m['Accuracy']:.4f}  ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
m, mdl, yp, ypr = evaluate_model(
    'Random Forest',
    RandomForestClassifier(n_estimators=200, max_depth=12, min_samples_leaf=20,
                           class_weight='balanced', n_jobs=-1, random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m)
trained_models['Random Forest'] = (mdl, yp, ypr)
print(f"Random Forest       — Accuracy: {m['Accuracy']:.4f}  ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
m, mdl_gb, yp_gb, ypr_gb = evaluate_model(
    'Gradient Boosting',
    GradientBoostingClassifier(n_estimators=300, learning_rate=0.08,
                               max_depth=5, subsample=0.8, random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m)
trained_models['Gradient Boosting'] = (mdl_gb, yp_gb, ypr_gb)
print(f"Gradient Boosting   — Accuracy: {m['Accuracy']:.4f}  ROC-AUC: {m['ROC-AUC']:.4f}")
```

```python
if XGBOOST_AVAILABLE:
    # scale_pos_weight = ratio of negative to positive samples — upweights the minority class
    scale_pw = float((y_train == 0).sum()) / float((y_train == 1).sum())
    m, mdl_xgb, yp_xgb, ypr_xgb = evaluate_model(
        'XGBoost',
        XGBClassifier(
            n_estimators=300, learning_rate=0.08, max_depth=6,
            subsample=0.8, colsample_bytree=0.8,
            scale_pos_weight=scale_pw,
            eval_metric='logloss', n_jobs=-1, random_state=42, verbosity=0
        ),
        X_train_f, X_test_f, y_train, y_test
    )
    all_results.append(m)
    trained_models['XGBoost'] = (mdl_xgb, yp_xgb, ypr_xgb)
    print(f"XGBoost             — Accuracy: {m['Accuracy']:.4f}  ROC-AUC: {m['ROC-AUC']:.4f}")
else:
    print('XGBoost not available — restart kernel after installing.')
```

```python
# Summary comparison table
results_df = pd.DataFrame(all_results).set_index('Model').sort_values('ROC-AUC', ascending=False)

STATS['best_model']    = results_df.index[0]
STATS['best_roc_auc']  = results_df['ROC-AUC'].iloc[0]
STATS['best_accuracy'] = results_df['Accuracy'].iloc[0]
STATS['best_recall']   = results_df['Recall'].iloc[0]
STATS['best_f1']       = results_df['F1 Score'].iloc[0]

print('Model Comparison (21 clean, leak-free features):')
print(results_df.round(4).to_string())
print(f'\nBest model: {STATS["best_model"]}  |  ROC-AUC: {STATS["best_roc_auc"]:.4f}')

res_long = results_df.reset_index().melt(id_vars='Model', var_name='Metric', value_name='Score')
fig = px.bar(res_long, x='Metric', y='Score', color='Model', barmode='group', range_y=[0.45, 1.0],
             title='Model Performance Comparison',
             text=res_long['Score'].round(3))
fig.update_traces(textposition='outside', textfont_size=9)
fig.update_layout(height=500)
fig.show()
```

<a id='8'></a>
## 8. Threshold Tuning

XGBoost uses `scale_pos_weight` to upweight cancellation predictions, so it already achieves strong recall at the default threshold of 0.5. Threshold tuning lets operators adjust the precision-recall tradeoff based on their business priorities.

| Threshold | Recall | Precision | Use Case |
|-----------|--------|-----------|----------|
| 0.25–0.35 | > 0.93 | Low | Aggressive revenue protection |
| 0.50 (default) | 0.813 | 0.588 | Natural XGBoost operating point |
| **0.55 (recommended)** | **0.770** | **0.625** | **Best F1 — recommended for deployment** |
| 0.65–0.75 | ~0.60 | > 0.70 | Luxury hotels with strict overbooking limits |

> In production, the threshold should be selected via cross-validation on the training set — not on the test set.

```python
best_prob  = trained_models[STATS['best_model']][2]
thresholds = np.arange(0.15, 0.80, 0.05)
thr_rows   = []

for thr in thresholds:
    yp_thr = (best_prob >= thr).astype(int)
    thr_rows.append({
        'Threshold' : round(float(thr), 2),
        'Accuracy'  : accuracy_score(y_test, yp_thr),
        'Precision' : precision_score(y_test, yp_thr, zero_division=0),
        'Recall'    : recall_score(y_test, yp_thr, zero_division=0),
        'F1 Score'  : f1_score(y_test, yp_thr, zero_division=0),
    })

thr_df      = pd.DataFrame(thr_rows)
best_f1_thr = thr_df.loc[thr_df['F1 Score'].idxmax(), 'Threshold']
STATS['recommended_threshold'] = best_f1_thr

print(thr_df.round(3).to_string(index=False))
print(f'\nBest F1 threshold: {best_f1_thr}')
```

```python
fig = go.Figure()
for metric, color in [('Recall', '#e74c3c'), ('Precision', '#3498db'),
                       ('F1 Score', '#2ecc71'), ('Accuracy', '#f39c12')]:
    fig.add_trace(go.Scatter(x=thr_df['Threshold'], y=thr_df[metric],
                             mode='lines+markers', name=metric, line=dict(color=color, width=2)))

fig.add_vline(x=0.5,         line_dash='dash', line_color='gray',
              annotation_text='Default (0.5)', annotation_position='top right')
fig.add_vline(x=best_f1_thr, line_dash='dot',  line_color='red',
              annotation_text=f'Best F1 ({best_f1_thr})', annotation_position='top left')
fig.update_layout(title=f'Threshold Tuning — {STATS["best_model"]}',
                  xaxis_title='Decision Threshold', yaxis_title='Score', height=450)
fig.show()

print(f'Classification report at recommended threshold = {best_f1_thr}:')
print(classification_report(y_test, (best_prob >= best_f1_thr).astype(int),
                             target_names=['Not Canceled', 'Canceled']))
```

## 9. Deposit Type Deep Dive

`deposit_type` is included despite the Non Refund recording artefact. This section quantifies how much the model actually relies on it by comparing Gradient Boosting with and without the feature.  
A ROC-AUC drop below 3% confirms that the remaining 20 features carry genuine, independent predictive signal.

```python
m_gb_nd, mdl_gb_nd, yp_gb_nd, ypr_gb_nd = evaluate_model(
    'GB (no deposit_type)',
    GradientBoostingClassifier(n_estimators=300, learning_rate=0.08,
                               max_depth=5, subsample=0.8, random_state=42),
    X_train_nd, X_test_nd, y_train, y_test
)

m_gb_row = next(r for r in all_results if r['Model'] == 'Gradient Boosting')
comp_df  = pd.DataFrame([m_gb_row, m_gb_nd]).set_index('Model')
roc_drop = m_gb_row['ROC-AUC'] - m_gb_nd['ROC-AUC']
STATS['deposit_roc_drop'] = roc_drop

print(comp_df.round(4).to_string())
print(f'\nROC-AUC drop when deposit_type is removed: {roc_drop:.4f}')

if roc_drop < 0.03:
    print('Drop < 3% — model is not over-relying on deposit_type. The other features carry independent signal.')
elif roc_drop < 0.05:
    print('Moderate drop (3–5%) — deposit_type adds signal but model remains robust.')
else:
    print('Drop > 5% — model relies heavily on deposit_type. Investigate further.')

comp_long = comp_df.reset_index().melt(id_vars='Model', var_name='Metric', value_name='Score')
fig = px.bar(comp_long, x='Metric', y='Score', color='Model', barmode='group', range_y=[0.4, 1.0],
             title='Gradient Boosting: With vs Without deposit_type',
             text=comp_long['Score'].round(3))
fig.update_traces(textposition='outside', textfont_size=9)
fig.show()
```

## 10. Feature Importance & Cross-Validation

Two complementary views of feature importance:

| Measure | Method | Top Feature |
|---------|--------|-------------|
| GB Feature Importance | Average split-quality gain across all trees | `lead_time` (0.195) |
| SHAP Mean \|SHAP\| | Average absolute prediction shift per booking | `country_group` (0.748) |

The difference is expected — SHAP captures *how much* a feature moves the prediction for each booking, while feature importance measures *how often* it is used for splitting. Both approaches consistently identify `lead_time`, `country_group`, `total_of_special_requests`, and `market_segment` as the top predictors.

```python
# Feature importance from the Gradient Boosting model trained with deposit_type
fi_full = pd.DataFrame({'Feature': FEATURES_FULL, 'Importance': mdl_gb.feature_importances_})
fi_full = fi_full.sort_values('Importance', ascending=False)

dep_rank = list(fi_full['Feature']).index('deposit_type') + 1
STATS['top_feature']  = f"{fi_full.iloc[0]['Feature']} ({fi_full.iloc[0]['Importance']:.3f})"
STATS['deposit_rank'] = dep_rank

print(fi_full.to_string(index=False))
print(f"\nTop feature  : {fi_full.iloc[0]['Feature']}  ({fi_full.iloc[0]['Importance']:.4f})")
print(f'deposit_type : rank #{dep_rank} of {len(fi_full)}')

fig = px.bar(fi_full, x='Importance', y='Feature', orientation='h',
             title='Feature Importances — Gradient Boosting (with deposit_type)',
             color='Importance', color_continuous_scale='Blues_r')
fig.update_layout(yaxis={'categoryorder': 'total ascending'}, height=650)
fig.show()
```

```python
# Feature importance without deposit_type — shows the true business-driven predictors
fi_nd = pd.DataFrame({'Feature': FEATURES_NO_DEP, 'Importance': mdl_gb_nd.feature_importances_})
fi_nd = fi_nd.sort_values('Importance', ascending=False)
STATS['top_features_no_deposit'] = fi_nd['Feature'].head(6).tolist()

print(fi_nd.to_string(index=False))
print('\nThe top-5 predictors are consistent whether or not deposit_type is included.')

fig = px.bar(fi_nd, x='Importance', y='Feature', orientation='h',
             title='Feature Importances — Gradient Boosting (without deposit_type)',
             color='Importance', color_continuous_scale='Oranges_r')
fig.update_layout(yaxis={'categoryorder': 'total ascending'}, height=650)
fig.show()
```

```python
# 5-Fold Stratified Cross-Validation
# Using n_estimators=150 (vs 300 in the main model) to keep runtime reasonable
cv_model  = GradientBoostingClassifier(n_estimators=150, learning_rate=0.08,
                                        max_depth=5, subsample=0.8, random_state=42)
skf       = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(cv_model, X_full, y, cv=skf, scoring='roc_auc', n_jobs=-1)

STATS['cv_mean'] = cv_scores.mean()
STATS['cv_std']  = cv_scores.std()
gap = STATS['best_roc_auc'] - cv_scores.mean()

print(f'ROC-AUC per fold: {cv_scores.round(4)}')
print(f'Mean ± Std      : {cv_scores.mean():.4f} ± {cv_scores.std():.4f}')
print(f'Test AUC        : {STATS["best_roc_auc"]:.4f}  (gap vs CV mean: {gap:.4f})')

if cv_scores.std() < 0.01:
    print('Model is very stable across folds (std < 0.01).')
elif cv_scores.std() < 0.02:
    print('Model is stable across folds.')
else:
    print('Some variance across folds — consider additional regularisation.')

if abs(gap) <= 0.015:
    print('Test AUC is consistent with CV mean (gap ≤ 0.015).')
else:
    print('Gap > 0.015 — review train/test split.')

fig = px.bar(x=[f'Fold {i+1}' for i in range(5)], y=cv_scores,
             text=cv_scores.round(4),
             title='5-Fold Cross-Validation ROC-AUC',
             labels={'x': 'Fold', 'y': 'ROC-AUC'},
             color=cv_scores, color_continuous_scale='Blues')
fig.add_hline(y=cv_scores.mean(), line_dash='dash', line_color='red',
              annotation_text=f'Mean = {cv_scores.mean():.4f}')
fig.update_traces(texttemplate='%{text}', textposition='outside')
fig.show()
```

<a id='11'></a>
## 11. SHAP Value Analysis

SHAP (SHapley Additive exPlanations) assigns each feature a contribution score for every individual prediction, capturing the *actual magnitude* of its impact on the model output.

SHAP is computed on the **Gradient Boosting** model. The saved `shap_explainer.pkl` (Section 12) is also for this model — it powers per-booking explanations in the Streamlit app.

```python
if not SHAP_AVAILABLE:
    print('SHAP not available — run the install cell and restart the kernel.')
else:
    # Sample 2,000 test rows for SHAP (full computation is slow)
    rng        = np.random.RandomState(42)
    sample_idx = rng.choice(len(X_test_f), size=min(2000, len(X_test_f)), replace=False)
    X_shap     = X_test_f.iloc[sample_idx].copy()

    explainer   = shap.TreeExplainer(mdl_gb)
    shap_values = explainer.shap_values(X_shap)

    mean_shap = pd.DataFrame({
        'Feature'    : X_shap.columns,
        'Mean |SHAP|': np.abs(shap_values).mean(axis=0)
    }).sort_values('Mean |SHAP|', ascending=False)

    STATS['shap_top_feature'] = mean_shap.iloc[0]['Feature']
    print('Global feature importance via mean |SHAP| value:')
    print(mean_shap.to_string(index=False))
```

```python
if SHAP_AVAILABLE:
    # Beeswarm plot: red = high feature value, blue = low; right side = pushes toward cancellation
    plt.figure(figsize=(10, 8))
    shap.summary_plot(shap_values, X_shap, max_display=15, show=True)

    # Bar plot: mean absolute SHAP — global feature ranking
    plt.figure(figsize=(10, 6))
    shap.summary_plot(shap_values, X_shap, plot_type='bar', max_display=15, show=True)
```

```python
if SHAP_AVAILABLE:
    # Waterfall plot for a single high-risk booking
    best_prob_sample = trained_models[STATS['best_model']][2]
    best_prob_shap   = best_prob_sample[sample_idx]
    high_risk_mask   = best_prob_shap > 0.65
    idx = np.where(high_risk_mask)[0][0] if high_risk_mask.any() else 0

    print(f'Predicted cancel probability : {best_prob_shap[idx]:.3f}')
    print(f'Actual outcome               : {"Canceled" if y_test.iloc[sample_idx[idx]] == 1 else "Not Canceled"}')

    explanation = shap.Explanation(
        values       = shap_values[idx],
        base_values  = explainer.expected_value,
        data         = X_shap.iloc[idx].values,
        feature_names= list(X_shap.columns)
    )
    plt.figure(figsize=(10, 6))
    shap.waterfall_plot(explanation, max_display=12, show=True)
```
