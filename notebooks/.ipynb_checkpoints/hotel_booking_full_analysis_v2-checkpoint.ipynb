## 3. Data Cleaning & Preprocessing

```python
# ── Missing values audit ─────────────────────────────────────────────────────
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)
print("Columns with missing values:")
print(missing_df)

fig = px.bar(
    missing_df.reset_index(),
    x='index', y='Missing %',
    title='Missing Value Percentage by Column',
    labels={'index': 'Column', 'Missing %': 'Missing (%)'},
    color='Missing %',
    color_continuous_scale='reds'
)
fig.show()
```

```python
# ── Duplicate check ───────────────────────────────────────────────────────────
duplicates = df.duplicated().sum()
print(f"Duplicate rows detected: {duplicates:,}")
print(f"  ({duplicates/len(df)*100:.1f}% of data)")
print("  These will be removed in the cleaning step below.")
```

```python
# ── Full Data Cleaning ────────────────────────────────────────────────────────
df_clean = df.copy()
initial_rows = len(df_clean)

# Step 1: Drop duplicate rows
#   Rationale: 31,994 exact duplicates — likely system re-entries, not real bookings
df_clean.drop_duplicates(inplace=True)
print(f"Step 1 – Removed duplicates       : -{initial_rows - len(df_clean):,} rows (now {len(df_clean):,})")

# Step 2: Remove bookings with 0 guests (impossible / data error)
before = len(df_clean)
df_clean = df_clean[
    (df_clean['adults'] + df_clean['children'].fillna(0) + df_clean['babies']) > 0
]
print(f"Step 2 – Removed 0-guest bookings  : -{before - len(df_clean):,} rows (now {len(df_clean):,})")

# Step 3: Remove rows with negative ADR
before = len(df_clean)
df_clean = df_clean[df_clean['adr'] >= 0]
print(f"Step 3 – Removed negative ADR rows : -{before - len(df_clean):,} rows (now {len(df_clean):,})")

# Step 4: Drop 'company' (94% missing)
df_clean.drop(columns=['company'], inplace=True)

# Step 5: Drop data-leakage columns
#   'reservation_status' = direct indicator of cancellation AFTER the fact
#   'reservation_status_date' = date of that status change — also post-event
df_clean.drop(columns=['reservation_status', 'reservation_status_date'], inplace=True)
print(f"Step 4/5 – Dropped company + leakage columns")

# Step 6: Fill remaining missing values
df_clean['children'].fillna(0, inplace=True)
df_clean['country'].fillna('Unknown', inplace=True)
df_clean['agent'].fillna(0, inplace=True)

print(f"\n✅ Final clean dataset: {len(df_clean):,} rows × {df_clean.shape[1]} columns")
print(f"   Removed {initial_rows - len(df_clean):,} rows total ({(initial_rows-len(df_clean))/initial_rows*100:.1f}% of raw data)")
print(f"   Remaining missing values: {df_clean.isnull().sum().sum()}")
```

### 4.2 Lead Time Distribution vs Cancellation
**Insight:** Cancelled bookings had an average lead time of ~145 days vs ~80 days for kept bookings.  
The further in advance someone books, the more likely plans change — a critical predictor.

### 4.3 Deposit Type — Key Anomaly to Understand
**⚠️ Important Nuance:** Non Refund bookings show ~99.4% cancellation rate. This is NOT standard behaviour.  
In this dataset, non-refundable bookings are often recorded as 'Canceled' in the system even when the guest forfeits the deposit and the hotel keeps the money. The hotel still receives revenue — this is a **data recording artefact**, not a true business cancellation.  
This makes `deposit_type` an extremely powerful but potentially misleading predictor.

```python
deposit_cancel = df_clean.groupby('deposit_type')['is_canceled'].agg(['sum','count'])
deposit_cancel['cancel_rate'] = (deposit_cancel['sum']/deposit_cancel['count']*100).round(1)
deposit_cancel['n_bookings']  = deposit_cancel['count']
deposit_cancel = deposit_cancel.reset_index()
print(deposit_cancel[['deposit_type','n_bookings','sum','cancel_rate']].to_string(index=False))

fig = px.bar(
    deposit_cancel, x='deposit_type', y='cancel_rate', text='cancel_rate', color='deposit_type',
    title='Cancellation Rate by Deposit Type (%)',
    labels={'cancel_rate':'Cancellation Rate (%)', 'deposit_type':'Deposit Type'}
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

### 4.4 Cancellation by Market Segment
**Insight:** Groups (61%) and Online TA (37%) have the highest rates.  
Direct bookings (15%) and Complementary (12%) are the safest. This tells revenue teams to watch Group bookings closely.

```python
market_cancel = df_clean.groupby('market_segment')['is_canceled'].agg(['sum','count'])
market_cancel['cancel_rate'] = (market_cancel['sum']/market_cancel['count']*100).round(1)
market_cancel = market_cancel.sort_values('cancel_rate', ascending=False).reset_index()

fig = px.bar(
    market_cancel, x='market_segment', y='cancel_rate', text='cancel_rate',
    color='cancel_rate',
    title='Cancellation Rate by Market Segment (%)',
    labels={'cancel_rate':'Cancellation Rate (%)', 'market_segment':'Market Segment'},
    color_continuous_scale='RdYlGn_r'
)
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

### 4.6 Monthly Cancellation Trend  
**✅ Corrected Insight:** The peak cancellation months are **April and June (~41%)**, NOT July/August as previously stated.  
November has the lowest rate (~31%). Hotels should tighten overbooking policies heading into spring.

```python
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly = df_clean.groupby('arrival_date_month')['is_canceled'].agg(['sum','count'])
monthly['cancel_rate'] = (monthly['sum']/monthly['count']*100).round(1)
monthly = monthly.reindex(month_order).reset_index()

# Find true peak
peak_month = monthly.loc[monthly['cancel_rate'].idxmax(), 'arrival_date_month']
peak_rate  = monthly['cancel_rate'].max()
low_month  = monthly.loc[monthly['cancel_rate'].idxmin(), 'arrival_date_month']
low_rate   = monthly['cancel_rate'].min()
print(f"Peak cancellation : {peak_month} ({peak_rate}%)")
print(f"Lowest cancellation: {low_month} ({low_rate}%)")

fig = px.line(
    monthly, x='arrival_date_month', y='cancel_rate', markers=True,
    title='Monthly Cancellation Rate (%) — Peak is Spring, NOT Summer',
    labels={'arrival_date_month':'Month', 'cancel_rate':'Cancellation Rate (%)'}
)
fig.update_traces(line_color='#e74c3c', marker_color='#c0392b', marker_size=8)
fig.show()
```

## 5. Feature Engineering & ML Preparation

```python
df_ml = df_clean.copy()

# Derived features
df_ml['total_nights']      = df_ml['stays_in_weekend_nights'] + df_ml['stays_in_week_nights']
df_ml['total_guests']      = df_ml['adults'] + df_ml['children'] + df_ml['babies']
df_ml['has_agent']         = (df_ml['agent'] > 0).astype(int)
df_ml['room_type_changed'] = (df_ml['reserved_room_type'] != df_ml['assigned_room_type']).astype(int)
df_ml['is_weekend_only']   = (
    (df_ml['stays_in_weekend_nights'] > 0) & (df_ml['stays_in_week_nights'] == 0)
).astype(int)
df_ml['lead_time_bucket'] = pd.cut(
    df_ml['lead_time'],
    bins=[-1, 7, 30, 90, 180, 365, 9999],
    labels=['0-7d', '8-30d', '31-90d', '91-180d', '181-365d', '365+d']
)

# Value counts of new features for sanity check
print("room_type_changed distribution:")
print(df_ml['room_type_changed'].value_counts())
print("\nhas_agent distribution:")
print(df_ml['has_agent'].value_counts())
print("\nlead_time_bucket distribution:")
print(df_ml['lead_time_bucket'].value_counts().sort_index())
```

```python
# ── Feature sets ──────────────────────────────────────────────────────────────
# PRIMARY feature set — includes deposit_type
features_full = [
    'lead_time', 'total_nights', 'total_guests', 'adr',
    'previous_cancellations', 'previous_bookings_not_canceled',
    'booking_changes', 'days_in_waiting_list',
    'required_car_parking_spaces', 'total_of_special_requests',
    'is_repeated_guest', 'has_agent', 'room_type_changed', 'is_weekend_only',
    'hotel', 'meal', 'market_segment', 'distribution_channel',
    'deposit_type', 'customer_type', 'lead_time_bucket'
]

# SECONDARY feature set — excludes deposit_type
# Used to check whether model relies too heavily on the near-deterministic deposit_type
features_no_deposit = [f for f in features_full if f != 'deposit_type']

target = 'is_canceled'
cat_cols = ['hotel', 'meal', 'market_segment', 'distribution_channel',
            'deposit_type', 'customer_type', 'lead_time_bucket']

def prepare_X(df_in, feature_list, cat_columns):
    """Encode categorical columns and return X matrix."""
    df_out = df_in[feature_list].copy()
    df_out['lead_time_bucket'] = df_out['lead_time_bucket'].astype(str) if 'lead_time_bucket' in df_out.columns else None
    le = LabelEncoder()
    for col in cat_columns:
        if col in df_out.columns:
            df_out[col] = le.fit_transform(df_out[col].astype(str))
    return df_out

df_model_full       = df_ml[features_full + [target]].copy()
df_model_no_deposit = df_ml[features_no_deposit + [target]].copy()

# Encode both
X_full       = prepare_X(df_ml, features_full, cat_cols)
X_no_deposit = prepare_X(df_ml, features_no_deposit, [c for c in cat_cols if c != 'deposit_type'])
y = df_ml[target]

print(f"Full feature set      : {X_full.shape[1]} features")
print(f"Without deposit_type  : {X_no_deposit.shape[1]} features")
```

```python
# ── Train/Test split ──────────────────────────────────────────────────────────
X_train_f, X_test_f, y_train, y_test = train_test_split(
    X_full, y, test_size=0.2, random_state=42, stratify=y
)
X_train_nd, X_test_nd, _, _ = train_test_split(
    X_no_deposit, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Training rows : {X_train_f.shape[0]:,}")
print(f"Test rows     : {X_test_f.shape[0]:,}")
print(f"\nClass balance in test set:")
print(y_test.value_counts(normalize=True).rename({0:'Not Canceled', 1:'Canceled'}).round(3))
```

## 6. Model Training & Evaluation

```python
def evaluate_model(name, model, X_tr, X_te, y_tr, y_te):
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_te)
    y_prob = model.predict_proba(X_te)[:, 1]
    return {
        'Model'    : name,
        'Accuracy' : accuracy_score(y_te, y_pred),
        'Precision': precision_score(y_te, y_pred),
        'Recall'   : recall_score(y_te, y_pred),
        'F1 Score' : f1_score(y_te, y_pred),
        'ROC-AUC'  : roc_auc_score(y_te, y_prob),
    }, model, y_pred, y_prob

all_results    = []
trained_models = {}
```

```python
# Logistic Regression (needs scaling)
scaler      = StandardScaler()
X_train_sc  = scaler.fit_transform(X_train_f)
X_test_sc   = scaler.transform(X_test_f)

m_lr, model_lr, y_pred_lr, y_prob_lr = evaluate_model(
    'Logistic Regression', LogisticRegression(max_iter=1000, random_state=42),
    X_train_sc, X_test_sc, y_train, y_test
)
all_results.append(m_lr)
trained_models['Logistic Regression'] = (model_lr, y_pred_lr, y_prob_lr)
print("Logistic Regression done")
```

```python
# Decision Tree
m_dt, model_dt, y_pred_dt, y_prob_dt = evaluate_model(
    'Decision Tree',
    DecisionTreeClassifier(max_depth=8, min_samples_leaf=50, random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m_dt)
trained_models['Decision Tree'] = (model_dt, y_pred_dt, y_prob_dt)
print("Decision Tree done")
```

```python
# Random Forest
m_rf, model_rf, y_pred_rf, y_prob_rf = evaluate_model(
    'Random Forest',
    RandomForestClassifier(n_estimators=200, max_depth=12, min_samples_leaf=20,
                           n_jobs=-1, random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m_rf)
trained_models['Random Forest'] = (model_rf, y_pred_rf, y_prob_rf)
print("Random Forest done")
```

```python
# Gradient Boosting
m_gb, model_gb, y_pred_gb, y_prob_gb = evaluate_model(
    'Gradient Boosting',
    GradientBoostingClassifier(n_estimators=200, learning_rate=0.1,
                                max_depth=5, random_state=42),
    X_train_f, X_test_f, y_train, y_test
)
all_results.append(m_gb)
trained_models['Gradient Boosting'] = (model_gb, y_pred_gb, y_prob_gb)
print("Gradient Boosting done")
```

```python
if XGBOOST_AVAILABLE:
    m_xgb, model_xgb, y_pred_xgb, y_prob_xgb = evaluate_model(
        'XGBoost',
        XGBClassifier(n_estimators=200, learning_rate=0.1, max_depth=6,
                      eval_metric='logloss', n_jobs=-1, random_state=42),
        X_train_f, X_test_f, y_train, y_test
    )
    all_results.append(m_xgb)
    trained_models['XGBoost'] = (model_xgb, y_pred_xgb, y_prob_xgb)
    print("XGBoost done")
else:
    print("XGBoost skipped (not installed)")
```

```python
# ── Results table ─────────────────────────────────────────────────────────────
results_df = pd.DataFrame(all_results).set_index('Model')
results_df = results_df.sort_values('ROC-AUC', ascending=False)
print("Model Comparison (with deposit_type):")
print(results_df.round(4).to_string())
```

```python
results_long = results_df.reset_index().melt(id_vars='Model', var_name='Metric', value_name='Score')

fig = px.bar(
    results_long, x='Metric', y='Score', color='Model',
    barmode='group',
    title='Model Performance Comparison (with deposit_type)',
    text=results_long['Score'].round(3),
    range_y=[0.6, 1.0]
)
fig.update_traces(textposition='outside', textfont_size=9)
fig.update_layout(height=500)
fig.show()
```

```python
# ── ROC curves ────────────────────────────────────────────────────────────────
fig = go.Figure()
colors = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12', '#9b59b6']

for (name, (_, _, yp)), color in zip(trained_models.items(), colors):
    fpr, tpr, _ = roc_curve(y_test, yp)
    auc = roc_auc_score(y_test, yp)
    fig.add_trace(go.Scatter(
        x=fpr, y=tpr, name=f"{name} (AUC={auc:.3f})",
        line=dict(color=color, width=2)
    ))

fig.add_trace(go.Scatter(
    x=[0,1], y=[0,1], mode='lines', name='Random (AUC=0.500)',
    line=dict(color='gray', width=1, dash='dash')
))
fig.update_layout(
    title='ROC Curves – All Models',
    xaxis_title='False Positive Rate', yaxis_title='True Positive Rate', height=500
)
fig.show()
```

```python
# ── Confusion matrices ────────────────────────────────────────────────────────
n_models = len(trained_models)
ncols    = 2
nrows    = (n_models + 1) // ncols

fig, axes = plt.subplots(nrows, ncols, figsize=(14, 5 * nrows))
axes = axes.flatten()

for i, (name, (_, y_pred, _)) in enumerate(trained_models.items()):
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(
        cm, annot=True, fmt='d', ax=axes[i], cmap='Blues',
        xticklabels=['Not Canceled','Canceled'],
        yticklabels=['Not Canceled','Canceled']
    )
    tn, fp, fn, tp = cm.ravel()
    axes[i].set_title(
        f'{name}\nAcc={accuracy_score(y_test,y_pred):.3f} | '
        f'False Neg={fn:,} | False Pos={fp:,}'
    )
    axes[i].set_xlabel('Predicted')
    axes[i].set_ylabel('Actual')

for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.suptitle('Confusion Matrices – All Models', fontsize=14, y=1.01)
plt.show()

print("\n📝 Note on confusion matrix interpretation for hotel use:")
print("  False Negative = predicted 'Not Canceled' but guest actually canceled")
print("  → Hotel keeps the room blocked, loses revenue opportunity")
print("  False Positive = predicted 'Canceled' but guest actually shows up")
print("  → Hotel overbooks a room, potentially displaces a guest")
print("  For hotels: reducing False Negatives (higher Recall) is usually the priority.")
```

## 7. Threshold Tuning — Business-Relevant Recall vs Precision

The default threshold of 0.5 is not optimal for hotels. If the business priority is to **catch more cancellations** (so rooms can be re-sold), we can lower the threshold to increase Recall at the cost of some Precision.

```python
# Use the best model (Gradient Boosting)
best_name = results_df.index[0]
best_prob = trained_models[best_name][2]

print(f"Threshold tuning on: {best_name}\n")

thresholds     = np.arange(0.20, 0.75, 0.05)
threshold_rows = []

for thr in thresholds:
    y_pred_thr = (best_prob >= thr).astype(int)
    threshold_rows.append({
        'Threshold' : round(thr, 2),
        'Accuracy'  : accuracy_score(y_test, y_pred_thr),
        'Precision' : precision_score(y_test, y_pred_thr, zero_division=0),
        'Recall'    : recall_score(y_test, y_pred_thr, zero_division=0),
        'F1 Score'  : f1_score(y_test, y_pred_thr, zero_division=0),
    })

thr_df = pd.DataFrame(threshold_rows)
print(thr_df.round(3).to_string(index=False))
```

```python
fig = go.Figure()
for metric, color in [('Recall','#e74c3c'), ('Precision','#3498db'),
                       ('F1 Score','#2ecc71'), ('Accuracy','#f39c12')]:
    fig.add_trace(go.Scatter(
        x=thr_df['Threshold'], y=thr_df[metric],
        mode='lines+markers', name=metric, line=dict(color=color, width=2)
    ))

fig.add_vline(x=0.5, line_dash='dash', line_color='gray',
              annotation_text='Default (0.5)', annotation_position='top right')
fig.add_vline(x=0.35, line_dash='dot', line_color='red',
              annotation_text='High-Recall (0.35)', annotation_position='top left')

fig.update_layout(
    title=f'Threshold Tuning – {best_name} (Lower = Catch More Cancellations)',
    xaxis_title='Decision Threshold',
    yaxis_title='Score',
    height=450
)
fig.show()

# Show recommended threshold results
recommended_thr = 0.35
y_pred_recommended = (best_prob >= recommended_thr).astype(int)
print(f"\nRecommended threshold = {recommended_thr} (high-recall mode):")
print(classification_report(y_test, y_pred_recommended,
                             target_names=['Not Canceled', 'Canceled']))
```
